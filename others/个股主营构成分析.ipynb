```python
# -*- coding: utf-8 -*-
"""
Created on Mon Dec 28 22:32:51 2015

@author: ray
"""
%matplotlib inline
from WindPy import *
import pandas as pd
from sqlalchemy import create_engine
import datetime as dt

#w.stop()
w.start()

def get_factor(code,s_date,e_date):
    df = {}
    factor = ['sec_name','pe_ttm','pb_lf']
    #data = w.wsd("000001.SH", "pe_ttm,pb_lf", "2005-11-28", "2015-12-28", "Period=W;Fill=Previous") 
    data = w.wsd(code, factor, s_date, e_date, "Period=W;Fill=Previous")     
    if data.ErrorCode == 0:
        df = pd.DataFrame({data.Fields[0] : data.Data[0], data.Fields[1] : data.Data[1],data.Fields[2] : data.Data[2]},index=data.Times)
        df = df.dropna()
        df['code'] = code
        #print(df.describe(percentiles=[0.1,0.2,0.8,0.9]))
    return df 
def getAshares():
    data =  w.wset("SectorConstituent","date=20160722;sectorId=a001010100000000")
    df = pd.DataFrame(data.Data)
    dt = df.transpose()
    dt.columns = ['date','code','name']
    dt.index = data.Codes
    return dt
# -------------------------------------------------------------
print ("start data laoding ...")
ashares = getAshares()
stockList = list(ashares.code.values)
field_indu = "trade_code,sec_name,segment_sales,segment_industry_item,segment_industry_sales,segment_industry_cost,segment_industry_profit"
field_prod = "trade_code,sec_name,segment_sales,segment_product_item,segment_product_sales,segment_product_cost,segment_product_profit"
data_indu = w.wss(stockList,field_indu ,"rptDate=20151231;order=1")
data_prod = w.wss(stockList, field_prod ,"rptDate=20151231;order=1")
print( "load Finished")
```

```python
def majorAnaly(data):
    df = pd.DataFrame(data.Data)
    dt = df.transpose()
    dt.columns = ['证券代码','证券名称','major','主业名称','收入','成本','利润']
    dt.index = data.Codes
    dt['利润率'] = 100 * dt['利润'] / dt['收入']
    dt = dt.dropna()
    dt.major = dt.major.apply(lambda x: float(x.split(';')[0].split(':')[1].replace('%','')) )
    dt = dt[dt.major > 80.0]
    dt = dt.sort_values(by='利润率')
    dt = dt.tail(100)  #利润率500强
    dt
    # dt_gp = dt.groupby(['行业名称'])
    # dt_gp.count()
    return dt
majorAnaly(data_prod)
```
