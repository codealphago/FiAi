```python
import numpy as np
import datetime as dt
import pandas as pd
import seaborn
from matplotlib import pylab
import matplotlib.pyplot as plt
from matplotlib.pyplot import *
import sklearn
import talib
import sklearn.preprocessing
from lib import report
from lib import comm

pd.options.display.max_columns = 100
pd.options.display.max_rows = 300
```

```python
# 绘制新股水位线
# 新股水位 = 平均开板价格与上市价格的比值
# 当前水位 = 当前价格与上市价格的比值

def newissueLevel(edate,windows=365):
    newLevel = pd.DataFrame()
    newSec_list = comm.getNewIssues(trade_date = edate , days=windows)   
    hqAll = DataAPI.MktEqudAdjGet(ticker=newSec_list.ticker,endDate=edate,field=u"ticker,tradeDate,secShortName,closePrice,preClosePrice,turnoverVol,turnoverValue,turnoverRate")
    for i in newSec_list.ticker :
        hq = hqAll[hqAll.ticker == i ]
        hq = hq.reset_index()
        #计算总换手率，交易天数,当天平均价格，平均成本
        hq['total_turnRate'] = hq.turnoverRate.cumsum() 
        hq['days'] = hq.index +1
        hq['avg_price'] =   hq.turnoverValue / hq.turnoverVol
        hq['avg_cost'] =  0
        #hq.avg_cost.iloc[0] = hq.avg_price.iloc[0] * hq.turnoverRate.iloc[0] + hq.preClosePrice.iloc[0] *(1-hq.turnoverRate.iloc[0])
        #计算平均成本=本日平均价格*换手率+ 上日平均成本*（1-换手率）
        # for i in range(1,len(hq)):
        #      hq.avg_cost.iloc[i] = hq.avg_price.iloc[i] * hq.turnoverRate.iloc[i] + hq.avg_cost.iloc[i-1] *(1-hq.turnoverRate.iloc[i])
        #print hq        
        newHq = pd.DataFrame()
        issueHq = hq[hq.tradeDate==hq.tradeDate.min()]
        newHq = issueHq.loc[:,['ticker','secShortName']]
        newHq['issueDate'] = issueHq.tradeDate
        newHq['issuePrice'] = issueHq.preClosePrice
        lastHq = hq[hq.tradeDate==hq.tradeDate.max()]
        newHq['lastDate'] = lastHq.tradeDate.values
        newHq['lastPrice'] = lastHq.closePrice.values
        newHq['lastLevel'] =  lastHq.closePrice.values / issueHq.preClosePrice.values
        newHq['days'] = lastHq.days.values

        hq = hq[(hq.total_turnRate > 0.5 ) ]
        if hq.empty == False:  
            openHq = hq[hq.total_turnRate == hq.total_turnRate.min()]
            newHq['openDate'] = openHq.tradeDate.values
            newHq['openPrice'] = openHq.closePrice.values
            newHq['OpenLevel'] = openHq.closePrice.values/ issueHq.preClosePrice.values
            newHq['days'] = openHq.days.values
        else : 
            newHq['openDate'] = np.nan
            newHq['OpenLevel'] = newHq.lastLevel
        newLevel = newLevel.append(newHq)
    newLevel.index = newLevel.ticker
    newSec_list = newSec_list.loc[:,['nonrestFloatShares','totalShares']]
    newLevel = pd.concat([newSec_list,newLevel],axis=1)   
    newLevel['totalAsset'] = newLevel.issuePrice *  newLevel.totalShares    / 10 ** 8
    newLevel['FloatAsset'] = newLevel.issuePrice *  newLevel.nonrestFloatShares / 10 **8      
    return newLevel

df = newissueLevel(dt.date.today()-dt.timedelta(1))
# 统计
df['issueMonth'] = df.issueDate.apply(lambda x : x[0:7])
#print df.to_html()
gp = df.groupby(['issueMonth'])
openlevel = gp.OpenLevel.mean()
opendays = gp.days.mean()
currlevel = gp.lastLevel.mean()

openlevel.plot(figsize=(16,4),kind='bar')
currlevel.plot(kind = 'line')
opendays.plot(kind='line')
axhline = openlevel[-1]

print "2016共上市:", len(df) ," 2016 发行规模:" ,  df.FloatAsset.sum()," 2016 发行总市值", df.totalAsset.sum() , "当前流通市值: ",sum(df.FloatAsset  * df.lastLevel), "当前总市值:",sum(df.totalAsset  * df.lastLevel)
gp.agg({'OpenLevel' : 'mean', 'totalAsset' : 'sum','FloatAsset' : 'sum','days' : 'mean', 'lastLevel':'mean'})

```

```python
#绘制某只股票换手率行情图
def plotHq(hq):
    fig = plt.figure() # Create matplotlib figure
    ax = fig.add_subplot(111) # Create matplotlib axes
    ax2 = ax.twinx() # Create another axes that shares the same x-axis as ax.
    hq.plot(y=['total_turnRate'],ax=ax, figsize=(18, 4), linewidth='2',color='k')
    ax.axhline(y=2.5, color='k')
    ax.axhline(y=3, color='k')
    ax.axhline(y=5, color='k')
    report.fig_style(ax, [u'总换手'],'upper left')

    ax2 = hq.plot(y=['closePrice','avg_cost'] ,ax=ax2, figsize=(18, 4), title= unicode(hq.secShortName.values[0],"utf-8") + hq.ticker.values[0], kind='line',color='rb')
    report.fig_style(ax2, ['Curr_price','avg_cost'],'upper right')
```

```python
def new_issue_stetage1():
    seclist = comm.getNewIssues(days=80)
    buys = pd.DataFrame()
    sells = pd.DataFrame()
    stats = pd.DataFrame()
    for i in seclist.index:
        sec = seclist.loc[i]
        hq,new =  comm.statNewIssue(sec)
        new = new.reset_index(drop=True)
        stats = stats.append(new,ignore_index=True)
        buy = hq[(hq.total_turnRate>3) & (hq.total_turnRate<5.0) & ( (hq.PE < 120) | (hq.ROE > 0.15) ) & (hq.netMarketVal < 60 )  ].head(1)
        buys = buys.append(buy,ignore_index=True )
        if len(buy) ==1: 
            sell = hq[ (hq.total_turnRate>5.0) | ((hq.closePrice > buy.closePrice.values[0] * 1.08 ) & (hq.tradeDate > buy.tradeDate.values[0]))  ].head(1)
            if len(sell) == 0 : 
                sell = hq.tail(1)
                sell.days = None
            sells = sells.append(sell,ignore_index=True )

    #买入卖出时间
    buyTickers = buys.loc[:,['ticker','closePrice','tradeDate','secShortName','ROE','PE','days','netMarketVal','totalMarketVal']]
    buyTickers.index = buyTickers.ticker
    sellTickers = sells.loc[:,['ticker','closePrice','tradeDate','secShortName','days']]
    sellTickers.columns = ['s_ticker','s_closePrice','s_tradeDate','s_secShortName','s_days']

    sellTickers.index = sellTickers.s_ticker
    trade1 = pd.concat([buyTickers,sellTickers],axis=1)
    trade1['ret'] = 100.0*(trade1.s_closePrice/trade1.closePrice -1)
    trade1['during'] = trade1.s_days - trade1.days
    return trade1

trade1 = new_issue_stetage1()
print '实现回报： ',np.mean(trade1[trade1.s_days > 0].ret) , "未实现回报", np.mean(trade1[trade1.s_days.isnull()].ret)
print trade1.sort(['s_tradeDate']).to_html()
```

```python
def new_issue_stetage2():
    #策略2:  开板日跌25% 买入, 涨10% 卖出.
    seclist = comm.getNewIssues(days=100)
    buys = pd.DataFrame()
    sells = pd.DataFrame()
    stats = pd.DataFrame()
    for i in range(len(seclist)):
        sec = seclist.iloc[i]
        hq,new =  comm.statNewIssue(sec)
        hq0 = new[new.flag == '开板日']
        if len(hq0) ==1: 
            openPrice = hq0.closePrice.values[0] 
            buy = hq[(hq.tradeDate > hq0.tradeDate.values[0]) & (hq.closePrice < openPrice * 0.75) & ( (hq.PE < 150) | (hq.ROE > 0.15)) & (hq.netMarketVal < 60 )].head(1)
            buys = buys.append(buy,ignore_index=True )
            if len(buy) ==1: 
                #print sec.ticker,buy.closePrice.values[0] , buy.tradeDate.values[0]
                sell = hq[(hq.closePrice > buy.closePrice.values[0] * 1.10 ) & (hq.tradeDate > buy.tradeDate.values[0])].head(1)
                if len(sell) == 0 : 
                    sell = hq.tail(1)
                    sell.days = None
                sells = sells.append(sell,ignore_index=True )  
    #买入卖出时间
    buyTickers = buys.loc[:,['ticker','closePrice','tradeDate','secShortName','ROE','PE','days','netMarketVal','totalMarketVal']]
    buyTickers.index = buyTickers.ticker
    sellTickers = sells.loc[:,['ticker','closePrice','tradeDate','secShortName','days']]
    sellTickers.columns = ['s_ticker','s_closePrice','s_tradeDate','s_secShortName','s_days']

    sellTickers.index = sellTickers.s_ticker
    trade2 = pd.concat([buyTickers,sellTickers],axis=1)
    trade2['ret'] = 100.0*(trade2.s_closePrice/trade2.closePrice -1)
    trade2['during'] = trade2.s_days - trade2.days
    return trade2

trade2 = new_issue_stetage2()
print '实现回报： ',np.mean(trade2[trade2.s_days > 0].ret) , "未实现回报", np.mean(trade2[trade2.s_days.isnull()].ret)
print trade2.sort(['netMarketVal']).to_html()
```
