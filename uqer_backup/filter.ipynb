```python
#coding=utf-8

import numpy as np
from CAL.PyCAL import *
from DataAPI import *
import datetime as dt

# 去除ST股（交易日不是S、ST、*ST、S*ST以及SST类上市公司股票；
def dropST(account,seclist):
    dt = Date.fromDateTime(account.current_date)
    cal = Calendar('China.SSE')
    last_day = cal.advanceDate(dt,'-1B',BizDayConvention.Preceding)            #计算出倒数第一个交易日
    last_day_str = last_day.strftime("%Y%m%d")
    try:
        STlist = DataAPI.SecSTGet(secID=seclist, beginDate=last_day_str, endDate=last_day_str, field=['secID']).tolist()
        seclist = [s for s in seclist if s not in STlist]
    except:
        pass
    return seclist
    
# 去除流动性差的股票(days 日内平均成交量>1000万)
def dropLiqid(account,seclist,days):
    tv = account.get_attribute_history('turnoverValue', days)
    mtv = {}
    for sec in tv.keys():
        mtv[sec] = np.average(tv[sec])
    seclist = [s for s in seclist if mtv[sec] >= 10000000]
    return seclist 

# 去除新上市或复牌的股票
def dropNew(account,seclist):
    opn = account.get_attribute_history('openPrice', 1)
    seclist = [s for s in seclist if not (np.isnan(opn.get(s, 0)[0]) or opn.get(s, 0)[0] == 0)]
    return seclist

# 去除弱势股票
def dropWeak(account,seclist,days):
    hist_prices = account.get_attribute_history('closePrice', days)
    hist_returns = {sec: hist_prices[sec][-1]/hist_prices[sec][0] for sec in hist_prices.keys()}
    seclist = [s for s in seclist if hist_returns.get(s, 0) > 0.96]
    return seclist

#---------------------------------------------------------------------------
# 去除ST股（交易日不是S、ST、*ST、S*ST以及SST类上市公司股票；
def removeST(seclist,trade_date=dt.date.today()):
    td = trade_date #- dt.timedelta(1)         
    tds =  td.strftime("%Y%m%d")
    STlist = DataAPI.SecSTGet(beginDate=tds, endDate=tds)
    #print STlist
    lst = list(STlist.secID)
    seclist = [s for s in seclist if s not in lst]
    return seclist

#print removeST(['002241.XSHE','600600.XSHG','000013.XSHE','002781.XSHE']) #move 000013
#print removeST(['002241','600600','000013','002781']) #move 000013

# 去除流动性差的股票(30 日内平均成交额>1000万)
def removeLiqid(seclist,trade_date=dt.datetime.today() ,days=30, threshold=10**7):
    
    td = trade_date - dt.timedelta(days)         
    begin =  td.strftime("%Y%m%d")
    end =  trade_date.strftime("%Y%m%d")
    mtv =  DataAPI.MktEqudGet(secID=seclist,beginDate=begin, endDate=end,field=['secID','turnoverValue'])
    #print mtv
    mtv = mtv.groupby('secID').mean()
    #print mtv
    mv = mtv[mtv.turnoverValue > threshold]
    #print mv
    seclist = list(mv.index)
    return seclist 

#print removeLiqid(['002241.XSHE','600600.XSHG','000711.XSHE','600779.XSHG']) #move 000711.XSHE
#-----------------------------------------------------------------------------------------------------
# 去除新上市或复牌的股票(上市小于90天股票,前天刚复牌股票)
def removeNew(seclist,trade_date=dt.date.today(),days=90):
    issuedate = trade_date - dt.timedelta(days)         
    issueDs =  issuedate.strftime("%Y-%m-%d")
    issue = DataAPI.EquGet(secID=seclist)
    #print issue
    iss = issue[issue.listDate <issueDs ]
    seclist = list(iss.secID)
    #move 前天刚复牌股票
    begin = trade_date - dt.timedelta(3)   
    begins = begin.strftime("%Y%m%d")
    ends = trade_date.strftime("%Y%m%d")
    halt  = DataAPI.SecHaltGet(secID=seclist,beginDate=begins,endDate = ends,field=['secID'])
    #print halt
    lst = list(halt.values)
    seclist = [s for s in seclist if s not in lst]
    return seclist 

#print removeNew(['002241.XSHE','600600.XSHG','000711.XSHE','600779.XSHG','002781.XSHE']) #move 002781.XSHE

# 去除弱势(下跌周期)股票(days 天 回报 > -0.04)
def removeWeak(seclist,trade_date=dt.date.today(),days=5,threshold=-0.04):
    begin = trade_date - dt.timedelta(days)   
    begins = begin.strftime("%Y%m%d")
    ends = trade_date.strftime("%Y%m%d")
    hist_prices  = DataAPI.MktEqudAdjGet(secID=seclist,beginDate=begins,endDate = ends,field=['secID','tradeDate','preClosePrice','closePrice'])
    hist_prices['ret'] = hist_prices.closePrice / hist_prices.preClosePrice -1
    hist_returns = hist_prices.groupby('secID').sum()
    #print(hist_returns)
    #print(hist_prices)
    seclist = [s for s in seclist if hist_returns.ret[s] > threshold]
    return seclist

#print removeWeak(['002241.XSHE','600600.XSHG','000711.XSHE','600779.XSHG','002781.XSHE']) #move 002781.XSHE,600600.XSHG
```
