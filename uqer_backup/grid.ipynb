```python
import pandas as pd
import lib.comm as comm

# 连续下跌天数
def continuous_loss_day(returnSeries):
    '''对于日期顺序排列的pd series，给出最近的连续下跌天数'''
    return len(returnSeries) - 1 - max([v for v in range(len(returnSeries)) if returnSeries[v] >= 0])


# 连续回撤
def continuous_drawdown(returnSeries):
    '''对于日期顺序排列的pd series，给出最近的连续回撤'''
    condrawdown = 1 - (1 + returnSeries[max([v for v in range(len(returnSeries)) if returnSeries[v] > 0]):][1:]).prod()
    condrawdown = 0 if np.isnan(condrawdown) else condrawdown
    return condrawdown


# 最大回撤
def max_drawdown(returnSeries, days_num):
    '''给出最近days_num个交易日内最大回撤'''
    valueSeries = (returnSeries + 1).cumprod()[-days_num:]
    return max([1 - valueSeries[x] / valueSeries[:x].max() for x in valueSeries.index])
    
    
# 前days_num日累计收益
def cumulative_return(returnSeries, days_num):
    return (1 + returnSeries[-days_num:]).prod() - 1


# （参考）现价与MA价比率
def price_to_MA(returnSeries, days_num):
    '''返回最新价格与days_num交易日MA价的比值'''
    valueSeries = (returnSeries + 1).cumprod()
    if days_num > len(returnSeries):
        print "MA窗口期超过上限，请调整到%i天以内！"%len(returnSeries)
    return (valueSeries[-1:] / valueSeries[-days_num:].mean())[0]


# 建仓至今股票累计收益率
def stock_return(account, symbol):
    try:
        return account.reference_price[symbol] / account.security_cost[symbol]
    except KeyError, e:
        print e,'not in valid_seccost / referencePrice.'

# 生成历史回报序列 sedID = '000006.XSHG' , 'benchmark' 
def get_his_return(account,secID= 'benchmark',his_windows = 20): 
    tempDate = account.get_symbol_history('tradeDate',his_windows)
    tradeDate = pd.DataFrame(tempDate).rename_axis({0:'tradeDate'}, axis = 1)
    if secID in account.universe + ['benchmark']: 
        tmpDict =  account.get_symbol_history(secID,his_windows)
        banchmark = pd.DataFrame(tmpDict).merge(tradeDate, left_index = True, right_index = True).set_index('tradeDate')
        banchmark['return'] = tmpDict['closePrice'] / tmpDict['preClosePrice'] - 1 
    else : 
        banchmark = tradeDate.set_index('tradeDate')
        banchmark['return'] = 0.00001 
    return  banchmark['return']

#buy4,buy3,buy2,buy1,sell1,sell2,sell3,sell4 = 0.88,0.91,0.94,0.97,1.03,1.06,1.09,1.12
def setup_grid(buy_size=0.03,sell_size=0.03):
    buy1 = 1 - buy_size 
    sell1 = 1 + sell_size 
    buy2 = 1 - 2 * buy_size 
    sell2 = 1 + 2 * sell_size 
    buy3 = 1 - 3 * buy_size 
    sell3 = 1 + 3 * sell_size 
    buy4 = 1 - 4 * buy_size 
    sell4 = 1 + 4 * sell_size 
    return buy4,buy3,buy2,buy1,sell1,sell2,sell3,sell4

# 根据网格调仓策略 
def adj_positon2(adj_pct=0,max_pct=0,buy_size=0.03, sell_size=0.03): 
    buy4,buy3,buy2,buy1,sell1,sell2,sell3,sell4 = setup_grid(buy_size ,sell_size )
    pct = -1  # 不调整
    if adj_pct < buy4:
        pct =  1.0 * max_pct
    elif adj_pct < buy3:
        pct =  0.9 * max_pct
    elif adj_pct < buy2:
        pct =  0.7 * max_pct
    elif adj_pct < buy1:
        pct =  0.6 * max_pct  
    elif adj_pct > sell4:
        pct =  0.0 * max_pct
    elif adj_pct > sell4:
        pct =  0.1 * max_pct
    elif adj_pct > sell2:
        pct =  0.3 * max_pct
    elif adj_pct > sell1:
        pct =  0.4 * max_pct
        
    return pct

def adj_positon(adj_pct=0,max_pct=0,buy_size=0.03, sell_size=0.03): 
    buy4,buy3,buy2,buy1,sell1,sell2,sell3,sell4 = setup_grid(buy_size ,sell_size )
    pct = -1 # 不调整
    if adj_pct < buy4:
        pct =  1.0 * max_pct
    elif adj_pct < buy3:
        pct =  0.9 * max_pct
    elif adj_pct < buy2:
        pct =  0.8 * max_pct
    elif adj_pct < buy1:
        pct =  0.7 * max_pct
    elif adj_pct < sell1:
        pct =  0.5 * max_pct
    elif adj_pct < sell2:
        pct =  0.4 * max_pct
    elif adj_pct < sell3:
        pct = 0.3 * max_pct
    elif adj_pct < sell4:
        pct =  0.2 * max_pct
    else : 
        pct = 0.0
    return pct
        
# 自定义选股函数
def stock_filter(tradeDate, seclist,stock_num):
    data = DataAPI.MktStockFactorsOneDayGet(tradeDate, seclist, field = 'secID,tradeDate,PE,LFLO,ROE')
    secs = list(data[(data.PE < 60)  & (data.PE > 0)  & (data.ROE > 0.07) ].sort('LFLO')[:30].secID)
    #secs = list(data[(data.PE < 50) & (data.ROE > 0.15)].sort('LFLO')[:30].secID)
    if len(secs) > 0:
        beginDate = comm.getPrevBusinessDay(tradeDate,250)
        histDf = DataAPI.MktEqudAdjGet(secID = secs, beginDate = beginDate, endDate = tradeDate, field = 'secID,tradeDate,preClosePrice,closePrice')
        histDf['return'] = histDf.closePrice / histDf.preClosePrice - 1 
        voldata = histDf.pivot(index = 'tradeDate', columns = 'secID', values = 'return').std(axis = 0).sort_values()[-stock_num:]
        secList = list(voldata.index)
        test = data[data.secID.isin(secList)].set_index('secID')
        features = test.merge(pd.DataFrame(voldata,columns=['vol']), left_index = True, right_index = True)
        #print features
    else : 
        secList, features= [],pd.DataFrame()
    return secList, features  
# 得到股票60日波动率
def stock_vol(tradeDate, banchmarkID ='000002.XSHE' ,  his_windows=250, vol_days=60): 
    beginDate = comm.getPrevBusinessDay(tradeDate,his_windows)
    histDf = DataAPI.MktEqudAdjGet(secID=banchmarkID,beginDate=beginDate,endDate=tradeDate,field="tradeDate,preClosePrice,closePrice",pandas="1")
    histDf['return'] = histDf.closePrice / histDf.preClosePrice - 1 
    vol = histDf['return'].std()
    vol60 = vol * np.sqrt(vol_days)
    return vol

# 得到市场60日波动率
def index_vol(tradeDate, banchmarkID ='399300' ,  his_windows=250,vol_days=60): 
    beginDate = comm.getPrevBusinessDay(tradeDate,his_windows)
    histDf = DataAPI.MktIdxdGet(ticker=banchmarkID,beginDate=beginDate,endDate=tradeDate,field="tradeDate,preCloseIndex,closeIndex",pandas="1")
    histDf['return'] = histDf.closeIndex / histDf.preCloseIndex - 1 
    vol = histDf['return'].std()
    vol60 = vol * np.sqrt(vol_days) 
    return vol
```
