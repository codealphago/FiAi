## 模型设置
我在对中国的A股市场做了测试，测试时间从2010年4月1日至2016年3月17日（包含牛熊市场），模型设置如下：
* 隐藏状态数（hidden state）：3；（涨，平，跌）
* 特征向量：基础指标（每日收盘价，每日资金流入，每日资金流出）构造出特征向量（日收益率，资金流入占当日所有流动资金比例，日总流动资金环比，（日总流动资金-过去一年平均日流动资金）/过去一年流动资金波动率
* 假设所有的特征向量的状态服从高斯分布（强假设，便于求解）

```python
from hmmlearn.hmm import GaussianHMM
import datetime
import numpy as np
from matplotlib import cm, pyplot as plt
import matplotlib.dates as dates
import pandas as pd
import seaborn as sns

sns.set_style('white')

pd.options.display.max_columns = 100
pd.options.display.max_rows = 300

import lib.beta as beta
import lib.alpha as alpha
import lib.valuation as valuation

color_map = {'LL':'r','LN':'y','SS':'b','SN':'g','NN':'k'}
label_map = {'LL':'^','LN':'^','SS':'.','SN':'.','NN':'*'}
```

```python
def training_model(X,status=3):
    # Make an HMM instance and execute fit
    model = GaussianHMM(n_components=status, covariance_type="diag", n_iter=1000).fit([X])
    statusFlag,detail,trans_matrix = analysis_status(model,X)
    return model,detail,statusFlag,trans_matrix
def calDuration(matrix):
    #print matrix,len(matrix),matrix.index[0],matrix.iloc[0,0]
    during = {}
    for i in range(len(matrix)):
        during[matrix.index[i]] = 1 / (1-matrix.iloc[i,i])
    return during
def analysis_status(model,X):
    statusFlag = {}
    detail = pd.DataFrame()
    var5d,var1d,mean1d,mean5d = [],[],[],[]

    for j in range(model.n_components):
        var5d.append(np.diag(model.covars_[j])[1])
        var1d.append(np.diag(model.covars_[j])[0])
        mean5d.append(model.means_[j][1])
        mean1d.append(model.means_[j][0])
    detail['mean1d'] = mean1d
    detail['mean5d'] = mean5d
    detail['var1d'] = var1d
    detail['var5d'] = var5d
    order_mean = np.sort(detail.mean1d)
    #print order_mean
    #牛市上涨LL，牛市震荡LN，熊市下跌SS，熊市震荡SN，未知NN
    for i in range(model.n_components):
        mean = model.means_[i]
        #print mean
        if  mean[0] == order_mean[-1]:  #牛市上涨
            statusFlag[i] = 'LL' 
        elif mean[0] == order_mean[0]:  #熊市下跌
            statusFlag[i] = 'SS' 
        else  :                                         #震荡市  
            statusFlag[i] = 'NN' 
    detail['status'] = statusFlag.keys()
    detail['label'] = statusFlag.values()
    
    trans_matrix = pd.DataFrame( model.transmat_,columns=detail.label.values,index=detail.label.values)
    #trans_matrix = pd.DataFrame(model.transmat_,columns=detail.label.values)
    return statusFlag,detail,trans_matrix

def status_map(status,statusFlag):
    st_map = []
    for s in status:
        st_map.append(statusFlag[s])
    return st_map

def model_quanlity(md,X,res,statusFlag,shift=-1):
    hidden_states = md.predict(X)
    hidden = status_map(hidden_states,statusFlag)
    hidden = np.array(hidden)
    status = res.indexStatus.values
    # print status
    # print hidden
    correct = 0.0 
    if shift == -1 or shift > len(hidden): 
        size = len(hidden)
    else : 
        size = shift
    for j in range(size):
        if hidden[-size + j] == status[-size + j]: 
            correct = correct + 1
    correct_rate = round(correct/size,2)
    
    return correct_rate
```

```python
def displayResult(hidden_states,res,statusFlag)  : 
    plt.figure(figsize=(15, 8))  
    for i in range(len(statusFlag)):
        idx = (hidden_states==i)
        flag = statusFlag[i] 
        plt.plot_date(res.index.values[idx],res.closeIndex.values[idx],label_map[flag[0:2]],label='%s state'%flag,lw=1 ,color=color_map[flag[0:2]]) 
        plt.legend()
        plt.grid(1)
        
def displaySigReturn(hidden_states,res,statusFlag, previous=True)  :
    plt.figure(figsize=(15, 8)) 
    for i in range(len(statusFlag)):
        idx = (hidden_states==i)
        flag = statusFlag[i] 
        if(previous == True):
            idx = np.append(0,idx[:-1])#获得状态结果后第二天进行买入操作
        #fast factor backtest
        df = res.Return1
        res['sig_ret%s'%i] = df.multiply(idx,axis=0)
        plt.plot(np.exp(res['sig_ret%s'%i].cumsum()),label='%s state'%flag ,color=color_map[flag[0:2]])
        plt.legend()
        plt.grid(1)
        
def displayReturn(hidden_states,res,statusFlag , strtegy = 'Long')  :
    idxlong = np.zeros(len(hidden_states))
    idxshort = np.zeros(len(hidden_states))
    for i in range(len(statusFlag)):
        flag = statusFlag[i] 
        if flag[0:1] == 'L' :
            idxlong = idxlong + (hidden_states==i)
        elif flag[0:1] == 'S' :
            idxshort = idxshort + (hidden_states==i)

    plt.figure(figsize=(15, 8))
    idxlong = np.append(0,idxlong[:-1])#获得状态结果后第二天进行买入操作 , 向后推一个位置
    idxshort = np.append(0,idxshort[:-1])#获得状态结果后第二天进行买入操作
    df = res.Return1
    
    if strtegy == 'Long' :
        res['sig_retbest'] =  df.multiply(idxlong,axis=0) 
    elif strtegy == "Short" :
        res['sig_retbest'] = - df.multiply(idxshort,axis=0)
    else :
        res['sig_retbest'] = df.multiply(idxlong,axis=0) - df.multiply(idxshort,axis=0)
    plt.plot_date(res.index.values,np.exp(res['sig_retbest'].cumsum()),'-',label= strtegy + '  strategy')
    plt.legend()
    plt.grid(1)
```

```python
# ticker 支持股票， beginDate 要比实际往前推250个交易日
def prepare_data(ticker, beginDate, endDate, hist_window=250):
    #使用 基础指标（每日收盘价，每日资金流入，每日资金流出）构造出特征向量（日收益率，资金流入占当日所有流动资金比例，日总流动资金环比，（日总流动资金-过去一年平均日流动资金）/过去一年流动资金波动率
    data = DataAPI.MktEqudAdjGet(ticker=ticker,beginDate=beginDate,endDate=endDate,field=['tradeDate','closePrice','lowestPrice','highestPrice','turnoverVol'])
    data1 = DataAPI.MktEquFlowGet(tradeDate=u"",secID=u"",ticker=ticker,beginDate=beginDate,endDate=endDate,field=u"secID,ticker,secShortName,tradeDate,moneyInflow,moneyOutflow,netMoneyInflow")
    data1 = data1.sort(['tradeDate'])
    
    #构造出特征向量（日收益率，资金流入占当日所有流动资金比例，日总流动资金环比，（日总流动资金-过去一年平均日流动资金）/过去一年流动资金波动率
    data['Return1'] = data.closePrice.pct_change()
    data['closeIndex']  = data.closePrice
    
    data1['totalFlow'] = data1.moneyInflow + data1.moneyOutflow
    data1['inflowPct']=data1.moneyInflow / data1.totalFlow 
    data1['totalPct'] = data1.totalFlow.pct_change()
    data1['yearMean'] = pd.rolling_mean(data1.totalFlow,hist_window)
    data1['yearVol'] = pd.rolling_std(data1.totalFlow,hist_window)
    data1['flowVol'] = (data1.totalFlow - data1.yearMean )/data1.yearVol
    
    
    res = pd.merge(data,data1,on='tradeDate')
    res = res.dropna()
    res = res.replace('inf',0.0)
    #res = res.set_index('tradeDate')
    
    #X = np.column_stack([res.Return1,res.inflowPct,res.totalPct,res.flowVol])
    X = np.column_stack([res.Return1,res.inflowPct,res.flowVol])
    
    #整合数据
    logReturn1 = res.Return1.values
    indexStatus = np.empty(len(logReturn1),dtype="S2")
    for i in range(len(indexStatus)):
        if logReturn1[i] >= 0.01 :
            indexStatus[i]= 'LL'
        elif logReturn1[i] <= -0.01 :
            indexStatus[i]= 'SS'     
        else :
            indexStatus[i]= 'NN'
    res['indexStatus'] = indexStatus     
    return X,res

# ticker = '000001'
# begindate = '20130401'
# enddate = '20150401'
# print  "train data from   [" ,begindate  , " to  " ,enddate , " ]"
# X,res = prepare_data(ticker,begindate,enddate)
# res
```
