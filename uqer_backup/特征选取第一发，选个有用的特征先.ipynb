```python
datafactor = DataAPI.MktStockFactorsDateRangeGet(secID=u"",ticker=u"000001",beginDate=u"20100101",endDate=u"20160801",field=u"",pandas="1")
dataday = DataAPI.MktEqudGet(tradeDate=u"",secID=u"",ticker=u"000001",beginDate=u"20100101",endDate=u"20160801",isOpen="1",field=u"",pandas="1")
data1 = pd.merge(datafactor,dataday,on = 'tradeDate')
data = data1.dropna(axis = 1)
data['adprice'] = data.closePrice*data.accumAdjFactor
data['zdf'] = 100*(data.preClosePrice-data.closePrice)/data.preClosePrice
```

选取000001这只股票，然后做一些数据上的预处理。

```python
##两段式分割函数
def ly(x):
    if x>0:
        a = 1
    else:
        a = 0
    return a
```

```python
##三段分割函数
def sy(x):
    if x>1:
        a = 2
    elif x<0:
        a = 1
    else:
        a = 0
    return a
```

```python
data['zd_'] = data['zdf'].apply(ly)
data['zd'] = data['zd_'].shift(-1)
data = data.dropna()
```

```python
###对所有变量进行划分
dummiesData={}
factordata = choosedata
t = 0
for factor in factorlistch:
    factorData=data.loc[:,factor]
    dummiesData[factor]= pd.qcut(factorData,4,labels = [1,2,3,4])
    factordata = pd.concat([factordata, dummiesData[factor]], axis=1)
    t = t+1
```

拿到一个懵逼的不会用的指标，我的第一反应是，如果，这个指标代表一些强弱势，那么，指标区间就可以分段，极强，强势，弱势，极弱。对一个指标分为了四段。所以，我就对所有指标做了四段分段。

```python
factordata
```

```python
###对每个特征打分
scoredata = []
for j in factordata.columns:
    factormean1 = abs(factordata['zd'][factordata[j] == 1].mean()-0.5)
    factormean2 = abs(factordata['zd'][factordata[j] == 2].mean()-0.5)
    factormean3 = abs(factordata['zd'][factordata[j] == 3].mean()-0.5)
    factormean4 = abs(factordata['zd'][factordata[j] == 4].mean()-0.5)
    absmean = factormean1+factormean2+factormean3+factormean4
    scoredata.append(absmean)
```

特征选取第一步就是删除一些无用变量。对变量打分，分数越高变量对其预测越是有用。

```python
###打分之后特征表
sfactor = pd.concat([pd.DataFrame(scoredata,columns = ['score']),pd.DataFrame(factordata.columns,columns = ['sfactor'])],axis = 1)
choosefactor = list(sfactor.sort(columns = 'score',ascending = False).sfactor[0:19])
```

特征的分数表，选取得分最高的20个变量

```python
choosefactor
```

选出打分前20的变量，进行onehot编码

```python
###通过字典方法的变量迭代 onehot编码
choosedict={}
chfactordata = choosedata
t = 0
for factor in choosefactor:
    factorData=data.loc[:,factor]
    choosedict[factor]= pd.get_dummies(pd.qcut(factorData,4,labels = [1,2,3,4]),prefix=factor)
    chfactordata = pd.concat([chfactordata, choosedict[factor]], axis=1)
    t = t+1
```

onehot的编码在于接下来利用特征选取的方法继续删除一些变量

```python
chfactordata
```

接下来就是机器学习部分，直接调用一些sklearn里的方法，简直不要好用！！！

```python
from sklearn.svm import LinearSVC
from sklearn.ensemble import ExtraTreesClassifier
from sklearn.feature_selection import SelectFromModel
from sklearn.ensemble import RandomForestClassifier
```

```python
x_train = chfactordata.iloc[0:int(0.75*len(factordata)),1:]
y_train = chfactordata.iloc[0:int(0.75*len(factordata)),0]
x_test = chfactordata.iloc[int(0.75*len(factordata)):,1:]
y_test = chfactordata.iloc[int(0.75*len(factordata)):,0]
```

选取前四分之三作为训练集，后四分之一作为测试集。

```python
clf = ExtraTreesClassifier()
clf.fit(x_train,y_train)
```

```python
clf.feature_importances_
```

```python
model = SelectFromModel(clf, prefit=True)
xtrain_new = model.transform(x_train)
xtest_new = model.transform(x_test)
```

```python
x_train.shape
```

```python
xtrain_new.shape
```

可以看到，进过特征提取之后，变量从81个下降到31个。

```python
prenew = []
for times in range(100):
    rfclf = RandomForestClassifier(n_estimators=10)
    rfclf = rfclf.fit(xtrain_new,y_train)
    pre = len(y_test[rfclf.predict(xtest_new) == y_test])
    prenew.append(pre)
newnum = pd.DataFrame(prenew).mean()
```

```python
preold = []
for times in range(100):
    rfclf = RandomForestClassifier(n_estimators=10)
    rfclf = rfclf.fit(x_train,y_train)
    pre = len(y_test[rfclf.predict(x_test) == y_test])
    preold.append(pre)
oldnum = pd.DataFrame(preold).mean()
```

```python
print oldnum/len(y_test),newnum/len(y_test)
```
