#迁移概率是指第一天涨幅change跳跃到第二天不同涨幅changejump的概率

#统计分析在后文，excel画出来明显一点
#得到的第一结论就是：
#无论哪个程度的change，第二个change基本围绕着0%，近似高斯的概率分布
#稍微有一点小差异出现在：当change>0是，changejump<0的概率稍微大于changejump>0的概率。change<0的时候类似。changejump是指第二天的change。.08以下则比较符合。
#极限的0.09则有点差异：第一天涨停的change=0.9，第二天changjump更可能大于零，更可能涨。0

#那么可以有这么一个策略：
#如果todaychange=0.01~0.08,卖入，很可能跌。todaychange=-0.01~-0.08，买入，很可能涨。RP太低遇上连跌就另外一回事了。风险起见，<0.09立刻卖，>0.10可以考虑追买


#采集所有股票的历史价格 统计今日涨幅和明日涨幅

```python
metadata=pd.DataFrame({'x':x,'y':y,'c':c})
metadata.to_csv('change_jumpchange_10,000.csv')
```
