# A1.短期风险: iVix 中国波动率指数 以及期权成交量指标
- 此处使用50ETF期权数据，计算上一交易日几项指标，追踪市场情绪变动；
- 初步讨论只包括两个指标 PCR 和 PCIVD (均为反向指标！！)，
- 另中国波指 iVIX 指数作为参考；细节可以参考 期权日报 和 期权择时指数 ；

## 图中名词解释如下：
- 反向指标 : PCR：Put Call Ratio 看跌看涨期权的成交量(成交额)比率
- 反向指标 : PCIVD：Put Call Implied Volatility Difference 看跌看涨期权隐含波动率差
- 中国波指 : iVIX： 用于衡量上证50ETF未来30日的预期波动 http://www.sse.com.cn/assortment/options/volatility/
---
1. iVix :正常范围[20 ,35]   , 上升看跌, 下降看涨,  > 35  继续上升(清仓); > 25 半仓 <20 继续下降  (满仓) , 上升就不加仓
2. PCR  :正常范围[0.5 , 1.2] , 上升看跌, 下降看涨,  > 1.2 继续上升(清仓); >1 半仓  < 0.5 继续下降  (满仓) , 上升就不加仓
---
### 今日预测:   20160418  有效期一周
<font color=#0099ff size=3 face="黑体">
- iVix :  23左右.  波动不大, 风险较小,连续两周下降趋势
- PCR  :  <0.5 :  看涨情绪上升, 可以加仓
</font>

````python
def optIndicators():
    trade_days = report.trade_pre_n_days(n=3) #两个月
    start, end = dt.datetime.strptime(trade_days[0], '%Y-%m-%d'), dt.datetime.strptime(trade_days[-1], '%Y-%m-%d')
  
    ### calculation
    opt_indicators = pd.DataFrame(columns=[u'成交额PCR',u'中国波指iVIX']) #,u'成交量PCR',u'近月PCIVD',u'次月PCIVD',])
    opt_indicators[[u'成交额PCR']] = report.histPCR50ETF(start, end)[[u'valuePCR']]
    # opt_indicators[u'中国波指iVIX'] = report.getHistDailyVIX(start, end).iVIX/100.0
    opt_indicators.index = [str(i)[:10] for i in opt_indicators.index.values]
    opt_indicators.index.name = 'tradeDate'
    
    # ### plot
    # ax = opt_indicators.plot(y=[u'中国波指iVIX'], figsize=(18, 4), title=u'中国波指iVIX 未来30日', linewidth='3')
    # axhline(y=0.15, color='g')
    # axhline(y=0.25, color='b')
    # axhline(y=0.35, color='r')
    # report.fig_style(ax, [u'中国波指iVIX'], 'best')

    ax = opt_indicators.plot(y=[ u'成交额PCR'], figsize=(18, 4), title=u'看跌期权成交额 / 看涨期权成交额', linewidth='3')
    axhline(y=0.5, color='g')
    axhline(y=1, color='b')
    axhline(y=1.5, color='r')
    report.fig_style(ax, [u'成交额PCR'], 'best')

    return opt_indicators

# data = optIndicators()
# data.tail(25)
````

# A2.中期风险: 融资融券余额/流通市值
## 融资融券余额
- 融资余额若增加时，表示投资者心态偏向买方，市场人气旺盛，属强势市场；反之则属于弱势市场。
- 融券余额指融券卖出额与每日偿还融券额的差额。融券余额增加，表示市场趋向卖方市场；反之则趋向买方。
## 融资融券余额/流通市值: 
- 从融资融券中可以看出人们对未来市场信心，人们是否敢于加杠杆。  
- 合理区间[2.5 4]: 小于2.5:信心严重不足, 大于4: 谨慎投资, 大于4.5: 信心过头, 目前中值 3 比较合适	
##  当前预测 
<font color=#0099ff size=3 face="黑体">
- 连续小额度回升,  当前信心在逐渐恢复中.  中期保持80%仓位参与.
</font>

## B2.观察: 大盘资金流入流出情况
- 目前是统计各个行业都资金流入流出情况来计算大盘资金流入流出情况，可以作为宏观择时的一个参考。


````python
def money_flow():
    """A股资金流入流出情况
    """
    trade_days = report.trade_pre_n_days(n=3)
    
    mkt = DataAPI.MktIdxdGet(ticker='000001',beginDate=trade_days[0],endDate=trade_days[-1],field=['tradeDate','closeIndex','turnoverValue'],pandas="1")
    mkt.index = mkt.tradeDate
    
    data = pd.DataFrame()
    for day in trade_days:
        tmp = DataAPI.MktIndustryFlowOrderGet(tradeDate=day)
        data = tmp if data.empty else data.append(tmp)
    data['Small']= (data.netInflowS + data.netInflowM ) / (10**8)
    data['Large'] = ( data.netInflowL + data.netInflowXl )/ (10**8)
    data['net'] = data.Small + data.Large
    data = data.groupby('tradeDate').sum()
    data = pd.concat([data,mkt],axis=1)
    data['rollavg10'] = pd.rolling_mean(data.net,10)
    data['rollavg22'] = pd.rolling_mean(data.net,22)
    data['diff'] = data.net - data.rollavg10
     
    fig = plt.figure() # Create matplotlib figure
    ax = fig.add_subplot(111) # Create matplotlib axes
    ax2 = ax.twinx() # Create another axes that shares the same x-axis as ax.
    
    data.plot(y=['rollavg22'],kind='line',ax= ax, figsize=(18, 5),title=u'每日资金净流入流出情况: (单位: 亿)',color='g')
    data.plot(y=['closeIndex'],kind='line',ax= ax2, figsize=(18, 5),color='r')
    report.fig_style(ax, [u'净流22日移动平均'] , 'upper left')
    report.fig_style(ax2, [ u'上涨指数'] )
    
    return data

data = money_flow()
data.loc[:,['Small','Large','net','closeIndex','rollavg10','rollavg22','diff']].tail(6)

# 如果当天流入大于5日平均 则买入时机， 如果小于卖出时机
````
