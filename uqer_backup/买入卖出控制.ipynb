```python
def stock_ma(tickerID):
    trade_days = report.trade_pre_n_d(n=140)
    df=DataAPI.MktEqudAdjGet(ticker=tickerID,beginDate=trade_days[0],endDate=trade_days[-1],field="tradeDate,lowestPrice,closePrice,turnoverValue,secShortName",pandas="1")
    lowest=np.array(df['lowestPrice'])
    vol=np.array(df['turnoverValue']).astype(float)
    mytitle = df.secShortName[0]
    #MA_Type: 0=SMA, 1=EMA, 2=WMA, 3=DEMA, 4=TEMA, 5=TRIMA, 6=KAMA, 7=MAMA, 8=T3 (Default=SMA)
    EMA20 = talib.MA(lowest,20,matype=0)
    VMA20 = talib.MA(vol,20,matype=1)
    # 20均线，量价齐升开始买入
    df['EMA20'] = EMA20
    df['VMA20'] = VMA20
    df['eFlag'] = df.lowestPrice - df.EMA20 
    df['vFlag'] = df.turnoverValue - df.VMA20
    df = df.dropna()
    df.index = df.tradeDate.values
    ax = df.plot(y=['lowestPrice','EMA20'], figsize=(18, 4),kind='line',title = unicode(mytitle, "utf-8") )
    report.fig_style(ax, [u'当日低价',u'价格均线'],'upper right')
    bx = df.plot(y=['turnoverValue','VMA20'], figsize=(18, 4),kind='line',title = unicode(mytitle, "utf-8") )
    report.fig_style(bx, [u'当日成交量',u'成交量均线'],'upper right') 
    return df

def statusBreak():
    trade_days = report.trade_pre_n_d(n=166)
    his_price = DataAPI.MktEqudAdjGet(ticker=ticker,beginDate=trade_days[0],endDate=trade_days[-1],field=u"tradeDate,closePrice,secShortName",pandas="1")
    his_price.index = his_price.tradeDate.values
    size = len(his_price)
    mytitle = his_price.secShortName.values[0]
    break_price = {}
    for i in range(45,size):
        price = his_price.closePrice.iloc[i-45:i]
        break_price[his_price.index[i]] =  comm.cal_updown_Status(price)
    df = pd.DataFrame(break_price) 
    df = df.transpose()
    df.columns = ['up_price','down_price','curr_price','status']
    ax = df.plot(y=['up_price','down_price','curr_price'], figsize=(18, 4),kind='line',title = unicode(mytitle, "utf-8"))
    report.fig_style(ax, [u'向上突破价',u'向下突破价',u'当前价'],'upper right')
    return df

def dazong_jiaoyi_detail(pool):
    """大宗交易
    """
    trade_days = report.trade_pre_n_days(n=1)
    print "大宗交易： 万元" , trade_days[-1]
    data = DataAPI.MktBlockdGet(ticker=pool, beginDate=trade_days[0], endDate=trade_days[-1], 
                                assetClass=u"E", field=['tradeDate', 'secID','secFullName','tradeVol','tradePrice','tradeVal',])
    if len(data) > 0 :
        data.tradeVol = data.tradeVol 
        grp = data.groupby(['tradeDate','secID','secFullName'])
        grp = grp.agg({'tradeVol' : 'sum','tradePrice' : 'mean','tradeVal':'sum'})
        return grp
    else :
        return pd.DataFrame()

def basis_report(quanlity,profitability):
    print "#1. 最近一季或多季收入增长25%以上, 或者至少最近3个季度收入百分比在加速"
    print "#2. 在公司所处行业最近一个季度税后利润名列前茅,或者接近行业新高"
    fig = plt.figure(1) 
    ax = fig.add_subplot(111) 
    ax12 = ax.twinx()
    quanlity.plot(y=['profit_rate'], figsize=(18, 4),ax=ax,kind='line')
    report.fig_style(ax, [u'利润率'],'upper left')
    quanlity.plot(y=['operateProfit_grow','revenue_grow'], ax=ax12, figsize=(18, 4),kind='bar', title=u'收入利润质量')
    report.fig_style(ax12, [u'利润增长同比',u'收入同比增长'],'upper right')

    print "#3. 最近一季增长至少25%"
    print "#4. 近3年年每股收益增长应该在25%以上"
    print "#5. 最近一个季度每股收益增长高一上一个季度的增长速度,即开始加速."
    fig2 = plt.figure(2) 
    ax2 = fig2.add_subplot(111)
    ax22 = ax2.twinx()
    quanlity.plot(y=['eps_grow'], ax=ax2, figsize=(18, 4),kind='line', title=u'收益质量',color='r')
    report.fig_style(ax2, [u'每股收益增长同比',],'upper left')
    quanlity.plot(y=['basicEPS'], figsize=(18, 4),ax=ax22, kind='bar')
    report.fig_style(ax22, [u'每股收益'],'upper right')   

    print profitability.tail(5).to_html()
    print "#6 ROE > 20 , ROA > 10"
    fig3 = plt.figure(3) 
    ax3 = fig3.add_subplot(111)
    ax33 = ax3.twinx()
    profitability.plot(y=['ROA','ROE'], ax=ax3, figsize=(18, 4),kind='bar', title=u'收益质量',color='gr')
    report.fig_style(ax3, [u'总资产收益率',u'净资产收益率',],'upper left')
    profitability.plot(y=['ROIC'], figsize=(18, 4),ax=ax33, kind='line')
    report.fig_style(ax33, [u'投入资本回报率'],'upper right')   
def stock_mark(tickerID):
    years = YearLocator()   # every year
    months = MonthLocator()  # every month
    yearsFmt = DateFormatter('%Y')

    color_map = {'LL':'r','LN':'y','SS':'b','SN':'g','XX':'k'}
    marker_map = {'LL':'^','LN':'*','SS':'s','SN':'d','XX':'.'}
    label_map = {'LL':'bull','LN':'rev','SS':'sell','SN':'SN','XX':'NOR'}
    
    trade_days = report.trade_pre_n_d(n=30)
    his =  DataAPI.MktEqudAdjGet(ticker=tickerID,beginDate=trade_days[0],endDate=trade_days[-1],field="tradeDate,highestPrice,lowestPrice,openPrice,preClosePrice,turnoverVol,turnoverValue,closePrice,ticker,secShortName",pandas="1")
    #his.index = his.tradeDate
    mytitle = his.secShortName.values[0]
    his['CHGPct'] = np.log(his.closePrice) - np.log(his.preClosePrice)
    his['volMean'] = pd.rolling_mean(his.turnoverVol,5)
    his['priceMean'] =  pd.rolling_mean(his.closePrice,5)
    his['CHGPctMean'] =  pd.rolling_mean(his.CHGPct,5)
    his= his.dropna()
    his = his.reset_index()
    size = len(his) 
    #初始标志   
    his['hidden_mark'] = 'XX'
    for x in range(1,size):
        pre = his.loc[x-1]
        curr = his.loc[x]
        volChg = abs(np.log(curr.turnoverVol) - np.log(pre.volMean))
        priceChg = np.log(curr.highestPrice) - np.log(curr.closePrice)   
        # 抛售标志(指数下跌 and 成交量 > 前日 and 成交量 > 前10日移动平均 )
        if curr.CHGPct < -0.002 and curr.turnoverVol > pre.turnoverVol  :#and curr.turnoverVol > curr.volMean:
            his.hidden_mark[x] = 'SS'
        # 滞涨行为(前期成交活跃,市场走高,之后遇到阻力,不能走高,上升幅度比不上前一两天)
        if abs(curr.CHGPct) < 0.003  and priceChg > 0.01 and volChg < 0.1 and curr.CHGPctMean > 0.005  :
            his.hidden_mark[x] = 'SN' 
        # 反弹信号
        if curr.CHGPct > 0.005 and curr.turnoverVol > pre.turnoverVol  and curr.turnoverVol > curr.volMean:
            his.hidden_mark[x] = 'LN'
        # 牛市信号(都是以一个反转日开始,大多数反转都发生在大盘试探性反弹的7天里的第4天-7天)
    # 绘图.
    plt.figure(figsize=(18, 4)) 
    for i in range(len(his)):
        flag =  his.hidden_mark[i] 
        plt.plot( his.index.values[i] , his.closePrice.values[i],marker =marker_map[flag],lw=1 ,color=color_map[flag]) 
        plt.legend()
        plt.grid(1) 
    # ax = his.plot(y=['closePrice'], figsize=(18, 4),title = unicode(mytitle, "utf-8"),colors='r',linestyle='',marker="d")
    # report.fig_style(ax, [u'收盘价'],'upper right')
    return his
```

```python
def money_flow_detail(pool):
    """资金流入流出情况
    """
    trade_days = report.trade_pre_n_days()
    print "资金流入流出： 万元" , trade_days[-1]
    data = pd.DataFrame()
    data = DataAPI.MktEquFlowOrderGet(ticker=pool,beginDate=trade_days[-2],endDate = trade_days[-1],field='tradeDate,secID,secShortName,netInflowS,netInflowM,netInflowL,netInflowXl')
    data.netInflowS = np.round(data.netInflowS / 10**4)
    data.netInflowM = np.round(data.netInflowM / 10**4)
    data.netInflowL = np.round(data.netInflowL / 10**4)
    data.netInflowXl = np.round(data.netInflowXl / 10**4)
    data['net'] = data.netInflowS + data.netInflowM +data.netInflowL +data.netInflowXl
    return data

df = money_flow_detail(ticker)
df.sort(['tradeDate','net'])
```
