`融资融券`

```python
def rong_zi_rong_quan_detail(pool = stocklist):
    """融资融券信息
    """
    trade_days = report.trade_pre_n_days()
    print u"融资融券： 万元" , trade_days[-12], trade_days[-7] , trade_days[-2] 
    data0 = DataAPI.FstDetailGet(ticker=pool, beginDate=trade_days[-2],endDate=trade_days[-2],field='tradeDate,secID,secShortName,tradeVal')
    data0.index = data0.secID
    data1 = DataAPI.FstDetailGet(ticker=pool, beginDate=trade_days[-7],endDate=trade_days[-7],field='tradeDate,secID,secShortName,tradeVal')
    data1.index = data1.secID
    data2 = DataAPI.FstDetailGet(ticker=pool, beginDate=trade_days[-12],endDate=trade_days[-12],field='tradeDate,secID,secShortName,tradeVal')
    data2.index = data2.secID
    data = data0
    data['change1W'] = data0.tradeVal - data1.tradeVal
    data['change2W'] = data0.tradeVal - data2.tradeVal
    data['pct_change%1W'] = 100*(data0.tradeVal / data1.tradeVal) -100
    data['pct_change%2W'] = 100*(data0.tradeVal / data2.tradeVal) -100

    data['change1W'] = np.round(data['change1W'] / 10**4,1)
    data['change2W'] = np.round(data['change2W'] / 10**4,1)
    data['tradeVal'] = np.round(data['tradeVal'] / 10**4,1)
    
    return data
df = rong_zi_rong_quan_detail(alist)
df
```

提示： 当两融余额属于收缩下降状态, 市场弱势,谨慎加仓

`大宗交易`

```python
def dazong_jiaoyi_detail(pool = stocklist):
    """大宗交易
    """
    trade_days = report.trade_pre_n_days(n=1)
    print "大宗交易： 万元" , trade_days[-1]
    data = DataAPI.MktBlockdGet(ticker=pool, beginDate=trade_days[0], endDate=trade_days[-1], 
                                assetClass=u"E", field=['tradeDate', 'secID','secFullName','tradeVol','tradePrice','tradeVal',])
    data.tradeVol = data.tradeVol 
    grp = data.groupby(['tradeDate','secID','secFullName'])
    grp = grp.agg({'tradeVol' : 'sum','tradePrice' : 'mean','tradeVal':'sum'})
    return grp

df = dazong_jiaoyi_detail(alist)
df
```

提示： 大宗交易比较平稳,每日小于一万手是安全区间.  留意大宗交易个股成交量和成交价

`资金流入流出情况`

```python
def money_flow_detail(pool = stocklist):
    """资金流入流出情况
    """
    trade_days = report.trade_pre_n_days()
    print "资金流入流出： 万元" , trade_days[-1]
    data = pd.DataFrame()
    data = DataAPI.MktEquFlowOrderGet(ticker=pool,beginDate=trade_days[-2],endDate = trade_days[-1],field='tradeDate,secID,secShortName,netInflowS,netInflowM,netInflowL,netInflowXl')
    data.netInflowS = np.round(data.netInflowS / 10**4)
    data.netInflowM = np.round(data.netInflowM / 10**4)
    data.netInflowL = np.round(data.netInflowL / 10**4)
    data.netInflowXl = np.round(data.netInflowXl / 10**4)
    data['net'] = data.netInflowS + data.netInflowM +data.netInflowL +data.netInflowXl
    
    return data

df = money_flow_detail(alist)
df.sort(['tradeDate','net'])
```
