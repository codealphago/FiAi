## Shibor的择时能力

（按：这是研究 market timing 的一个意外收获，完全没有想到会有如此的效果）

-----------------------

在传统的金融意义上，由于流动性的总量是一定的，而且风险偏好有着巨大的差异，一般认为股票市场和债券市场存在着跷跷板效应，股市和债市可以参考着看，有一定的借鉴意义。不过对于中国市场而言，股市和其他市场的关系一般没有那么紧密，所以利用Shibor来进行择时也只是姑且一试。

利用Shibor对股市进行择时的 intuition 即如同上述所说的，当利率处于高位继而下行时，流动性逐渐从货币市场和债券市场释放出来进入股市，会推高股指；反之，则意味着整个资本市场流动性降低，投资者逐渐转向低风险资产，

------------------------

策略细节方面，首先是Shibor期限的选取，作为择时信号显然要对市场变化敏感，故我选择了7天的Shibor作为信号源，因为其每天的波动过于剧烈，所以对其进行了三个月的平滑。

其次是对于Shibor高低位的判断，这次我尝试了分位点，即以过去120个交易日的Shibor作为经验分布，求出其一定的分位点作为阈值来和调仓日进行比较，我尝试了多种不同的分位点。

再然后是标的的选择，我尝试了沪深300、中证500和创业板指，代码示例中采用的是创业板指，因为表现实在是太抢眼了，为此我反复检查了三四遍确定了没有未来数据。

最后是多空的控制，虽然目标是高位看多低位看空，但还存在着中间情况的可能，对于这种情况，我选择直接看多，因为总是持有指数，即使表现不好，那么也是持仓与标的表现一样不好，在择时的意义上并没有额外的损失。

------------------------

```python
import numpy as np
import pandas as pd
from datetime import datetime
from lib import comm

IDXMAP = {'HS300': '000300.ZICN', 'ZZ500': '000905.ZICN', 'CYB': '399006.ZICN'}

def get_weekly_rebalance_days(start, end):
    df = DataAPI.TradeCalGet(exchangeCD="XSHG", beginDate=start, endDate=end, field="calendarDate,prevTradeDate,isOpen,isWeekEnd")
    df['currTradeDate'] = df['calendarDate']
    df['nextTradeDate'] = ''
    d = len(df)
    for i in range(d):
        for j in range(i+1, d):
            if df.at[j, 'isOpen']:
                df.at[i, 'nextTradeDate'] = df.at[j, 'currTradeDate']
                break
    df = df[df.isWeekEnd == 1].loc[:, ['currTradeDate', 'prevTradeDate', 'nextTradeDate']]
    return df

def get_index_info(indexID, start, end):
    index_df = DataAPI.MktIdxdGet(indexID=IDXMAP.get(indexID, indexID), beginDate=start, endDate=end, field='tradeDate,closeIndex')
    return index_df
```

```python
start = '2014-01-01'
end = '2016-12-31'
term = '1W'
shibor_name = 'shibor_{}'.format(term)
shibor_ma = '{}_ma'.format(shibor_name)
dist_window = 120  # 计算分布的时间窗口
ma_window = 60  # 计算平滑的时间窗口

# 处理分位点和平滑
shibor_df = comm.get_shibor(term, start, end)
shibor_df['1/3q'] = pd.rolling_quantile(shibor_df[shibor_name], dist_window, 0.33)
shibor_df['1/2q'] = pd.rolling_quantile(shibor_df[shibor_name], dist_window, 0.5)
shibor_df['2/3q'] = pd.rolling_quantile(shibor_df[shibor_name], dist_window, 0.67)
shibor_df[shibor_ma] = pd.rolling_mean(shibor_df[shibor_name], ma_window)

shibor_df = shibor_df[dist_window-1:]
shibor_df.reset_index(inplace=True, drop=True)

start = shibor_df.at[0, 'tradeDate']
week_df = get_weekly_rebalance_days(start, end)
index_df = get_index_info('CYB', start, end)
```

```python
shibor_df.plot(figsize=(16,5))
```

```python
def shibor_timing(shibor_df, index_df, week_df, criterion):
    summary_df = week_df.merge(shibor_df, how='left', left_on='currTradeDate', right_on='tradeDate')

    del summary_df['tradeDate']
    summary_df = summary_df.merge(index_df, how='left', left_on='nextTradeDate', right_on='tradeDate')
    del summary_df['tradeDate']
    summary_df['index_return'] = summary_df['closeIndex'] / summary_df['closeIndex'].shift(1)
    summary_df.at[0, 'index_return'] = 1.
    summary_df['idx_cum_ret'] = np.cumprod(summary_df['index_return'])

    summary_df['long_timing'] = summary_df[shibor_ma] >= summary_df[criterion]
    summary_df['short_timing'] = summary_df[shibor_ma] < summary_df[criterion]

    summary_df['long_cum_ret'] = 1.
    summary_df['short_cum_ret'] = 1.
    summary_df['longshort_cum_ret'] = 1.

    for i in range(1, len(summary_df)-1):
        summary_df.at[i+1, 'longshort_cum_ret'] = summary_df.at[i, 'longshort_cum_ret']
        
        summary_df.at[i+1, 'long_cum_ret'] = summary_df.at[i, 'long_cum_ret']
        if summary_df.at[i, 'long_timing']:
            summary_df.at[i+1, 'long_cum_ret'] *= summary_df.at[i+1, 'index_return']
            summary_df.at[i+1, 'longshort_cum_ret'] *= summary_df.at[i+1, 'index_return']
    
        summary_df.at[i+1, 'short_cum_ret'] = summary_df.at[i, 'short_cum_ret']
        if summary_df.at[i, 'short_timing']:
            summary_df.at[i+1, 'short_cum_ret'] *= 1. / summary_df.at[i+1, 'index_return']
            summary_df.at[i+1, 'longshort_cum_ret'] *= 1. / summary_df.at[i+1, 'index_return']
        
    return summary_df
```

```python
summary_df = shibor_timing(shibor_df, index_df, week_df, '1/2q')
summary_df.loc[:, ['idx_cum_ret', 'long_cum_ret', 'short_cum_ret', 'longshort_cum_ret']].plot(figsize=(16,5))
summary_df.loc[:, [shibor_ma, '1/3q', '1/2q', '2/3q']].plot(figsize=(16,5))
```
