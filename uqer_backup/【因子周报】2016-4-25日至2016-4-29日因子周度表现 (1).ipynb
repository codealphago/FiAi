`风格因子周度表现`
====
****

```
    目前优矿平台提供了近130个因子，我们依照 https://uqer.io/community/share/55e662f9f9f06c1ea481f9cf 中的方法，从各类风格因子中各选取3个分别构建Alpha模型，并检验了各风格模型在上周的表现：曲线代表各风格因子组合相对于标的的累计超额收益率。
```

`风格因子选取`

```
    盈利因子: ['ROA','ROE','EPS']
    规模因子: ['NetProfitRatio','DebtEquityRatio','FixAssetRatio']
    成长因子: ['NetAssetGrowRate', 'TotalAssetGrowRate', 'OperatingRevenueGrowRate']
    现金流因子: ['CashRateOfSales','FinancingCashGrowRate','OperCashGrowRate']
    营运能力因子: ['ARTRate','AccountsPayablesTRate','TotalAssetsTRate']
    分析师预期因子: ['REC','FY12P','SFY12P']
    估值类因子: ['PE','PS','PB']
```


````python
# -*- coding: utf-8 -*-
import numpy as np
import quartz as qz
import datetime as dt
import matplotlib.pyplot as plt
import matplotlib.pylab as mpb
from CAL.PyCAL import *
from matplotlib.font_manager import FontProperties
from matplotlib.ticker import MultipleLocator, FormatStrFormatter  
from matplotlib.dates import DayLocator, HourLocator, DateFormatter, drange



class charPtf(object):
    
    def __init__(self,portfolio,currDate,attribute,period = u''):
        ####the input of the portfolio
        self.ptfTicker = [index[:6] for index in portfolio]
        self.attrOrig = attribute
        self.periodOrig = period
        self.currDate = currDate
        
        ####the output of  the portfolio
        self.period = self.period()
        self.pdData = self.pdData()
        self.popKeys = self.popKeys()
        self.ptf = self.ptf()
        self.secID = [index for index in portfolio if index[:6] in self.ptf]
        self.attr = self.attr()
        self.e = np.ones((len(self.ptf),1))
        self.exsRet = self.exsRet(self.period)
        self.retMat = self.retMat()
        self.retCov = self.retCov()
        self.ptfAttr = self.ptfAttr()               #特征组合
        self.ptfAttrVar = self.ptfAttrVar()           #特征组合的方差
        
    def period(self):
        if self.period == u'':
            period = 100
        else:
            period = self.periodOrig
        return period
    
    
    def pdData(self):
        calendar = Calendar('China.SSE')
        endDate = calendar.advanceDate(self.currDate,'-'+str(1)+'B').toDateTime()
        beginDate = calendar.advanceDate(self.currDate,'-'+str(self.period)+'B').toDateTime()  
        
        ticker = self.ptfTicker
        pdData = {index:DataAPI.MktEqudGet(tradeDate=u"",secID=u"",ticker=index,beginDate=beginDate,endDate=endDate,field=[],pandas="1") for index in ticker}
        return pdData
    
    
    def popKeys(self):
        popKeys = [index for index in self.pdData if len(self.pdData[index])<self.period]
        return popKeys
        
    def ptf(self):
        return [index for index in self.ptfTicker if index not in self.popKeys]
    
    def attr(self):
        return np.mat([self.attrOrig[index] for index in self.ptf]).T
        
    def exsRet(self,period):
        ticker = self.ptf
        for index in self.popKeys:
            self.pdData.pop(index)
                
        exsRet = {}
        for index in ticker:
            exsRet[index] = (self.pdData[index]['closePrice'] - self.pdData[index]['preClosePrice'])/self.pdData[index]['preClosePrice']
        return exsRet
    
    def retMat(self):
        ticker = sorted(self.exsRet.keys(),key = lambda x:int(x))
        tmpList = []
        for index in ticker:
            tmpList.append(self.exsRet[index].tolist())
        return np.mat(tmpList)
        pass
    
    def retCov(self):
        try:
            return np.mat(np.cov(self.retMat))
        except Exception,e:
            print 'raise exception in return covariance!'
            pass
            
    def ptfAttr(self):
        try:
            ptfAttr = self.retCov.I*self.attr*(self.attr.T*self.retCov.I*self.attr).I                
            return ptfAttr
        except:
            return np.zeros(np.shape(self.attr))
        pass
    
    
    def ptfAttrVar(self):
        try:
            return self.ptfAttr.T*self.retCov*self.ptfAttr
        except:
            pass


class handleFactor(object):
    
    def __init__(self,currDate,factor,portfolio):
        ####Input
        self.currDate = self._toDatetime(currDate)
        self.factor = factor
        self.portfolio = portfolio
        
        ####Output
        self.yestDay = self.adjustDate()
        self.factorSig = self.factorSig()
        self.secID = self.secID()
    def _toDatetime(self,currDate):
        if type(currDate) == dt.datetime:
            return currDate
            pass
        elif type(currDate) == str:
            currDate = currDate.replace('-','')
            return dt.datetime(int(currDate[:4]),int(currDate[4:6]),int(currDate[6:8]))
        
    def adjustDate(self):
        cal = Calendar('China.SSE')
        period = Period('-1B')
        today = Date.fromDateTime(self.currDate)  # 向前移动一个工作日
        yesterday = cal.advanceDate(today, period)
        yesterday = yesterday.toDateTime()
        return yesterday

    def factorSig(self):
        factor = self.factor
        factorData = DataAPI.MktStockFactorsOneDayGet(tradeDate=self.yestDay.strftime('%Y%m%d'),secID=self.portfolio,field=["secID",factor],pandas="1")
        factorData.columns = ['secID',factor]
        factorData['ticker'] = factorData['secID'].apply(lambda x:x[:6])
        factorData.set_index('ticker',inplace = True)
        attr = factorData[factor].fillna(1).to_dict()
        secID = factorData['secID'].to_dict()
        signal = attr
        # signal = standardize(neutralize(winsorize(attr),self.yestDay.strftime('%Y%m%d')))
        return signal
    
    def secID(self):
        factor = self.factor
        factorData = DataAPI.MktStockFactorsOneDayGet(tradeDate=self.yestDay.strftime('%Y%m%d'),secID=self.portfolio,field=["secID",factor],pandas="1")
        factorData.columns = ['secID',factor]
        factorData['ticker'] = factorData['secID'].apply(lambda x:x[:6])
        factorData.set_index('ticker',inplace = True)
        attr = factorData[factor].fillna(1).to_dict()
        secID = factorData['secID'].to_dict()
        # signal = standardize(neutralize(winsorize(attr),self.yestDay.strftime('%Y%m%d')))
        return secID


def initialize(account):
    pass

def handle_data(account):
    portfolio = []
    factorList = []
    for factor in styleFactor:
        currSig = standardize(neutralize(winsorize(handleFactor(account.current_date,factor,account.universe).factorSig),account.previous_date.strftime('%Y%m%d')))
        if portfolio == []:
            portfolio = handleFactor(account.current_date,factor,account.universe).secID.values()
        factorList.append(currSig.values())
    
    signal = np.mat(factorList)
    allocator = np.mat([[1.0/len(styleFactor)]]*len(styleFactor))
    score = (signal.T*allocator).T.tolist()[0]
    
    scoreDict = dict(zip(portfolio,score))
    weightMap = simple_long_only(scoreDict,account.current_date.strftime('%Y%m%d'))
    asset = account.referencePortfolioValue
    
    # 计算调仓数量
    extension = {}
    for stock in account.universe:
        weight = weightMap.get(stock, 0)
        price = account.referencePrice[stock]
        if not np.isnan(weight):
            if not np.isnan(price):
                extension[stock] = int(asset * weight / price) - account.valid_secpos.get(stock, 0)

    # 按先卖后买的顺序发出指令
    for stock in sorted(extension, key=extension.get):
        order(stock, extension[stock])
        
        
#################################################################################
#   backtest parameters
#################################################################################
# start = '2016-04-15'                       # 回测起始时间
start = dt.datetime(2016,4,25)
end = dt.datetime(2016,4,29)
benchmark = 'HS300'                        # 策略参考标准
universe = set_universe('HS300')  # 证券池，支持股票和基金
capital_base = 100000                      # 起始资金
freq = 'd'                                 # 策略类型，'d'表示日间策略使用日线回测，'m'表示日内策略使用分钟线回测
refresh_rate = 5                           # 调仓频率，表示执行handle_data的时间间隔，若freq = 'd'时间间隔的单位为交易日，若freq = 'm'时间间隔为分钟


factors = {
    # 'profitability': ['ROA','ROE','EPS'],
    # 'capital': ['NetProfitRatio','DebtEquityRatio','FixAssetRatio'],
    # 'growth': ['NetAssetGrowRate', 'TotalAssetGrowRate', 'OperatingRevenueGrowRate'],
    # 'cash': ['CashRateOfSales','FinancingCashGrowRate','OperCashGrowRate'],
    # 'operating': ['ARTRate','AccountsPayablesTRate','TotalAssetsTRate'],
    # 'analyzer': ['REC','FY12P','SFY12P'],
    # 'assessment': ['PE','PS','PB']
    u'盈利因子': ['ROA','ROE','EPS'],
    u'规模因子': ['NetProfitRatio','DebtEquityRatio','FixAssetRatio'],
    u'成长因子': ['NetAssetGrowRate', 'TotalAssetGrowRate', 'OperatingRevenueGrowRate'],
    u'现金流因子': ['CashRateOfSales','FinancingCashGrowRate','OperCashGrowRate'],
    u'营运能力因子': ['ARTRate','AccountsPayablesTRate','TotalAssetsTRate'],
    u'分析师预期因子': ['REC','FY12P','SFY12P'],
    u'估值类因子': ['PE','PS','PB']
}

styleResult = {}
for style in factors:
    styleFactor = factors[style]
    simParam = qz.sim_condition.env.SimulationParameters(start, end, benchmark, universe, capital_base)
    idxmap_all, data_all = qz.sim_condition.data_generator.get_daily_data(simParam)
    strategy = qz.sim_condition.strategy.TradingStrategy(initialize, handle_data)
    bt,account = qz.quick_backtest(simParam, strategy, idxmap_all, data_all, refresh_rate = refresh_rate)
    perf = qz.perf_parse(bt, account)
    styleResult[style] = perf
    # print "The quick backtest on style <" + str(style) + "> has done!"

# print 'Backtest is done!'

````

````python
def style_factor_performance():
    #### Daily excess profit figure
    fontsize, labelsize, linewidth = 15, 15, 2
    font.set_size(22)
    titleFont = font.copy()
    titleFont.set_size(16)
    scaleFont = font.copy()
    scaleFont.set_size(16)
    legendFont = font.copy()
    legendFont.set_size(12)

    delta = dt.timedelta(hours=24)
    dates = drange(start, end + dt.timedelta(1), delta)
    
    
    xmajorLocator = MultipleLocator(1)  
    xmajorFormatter = DateFormatter('%Y-%m-%d')
    ymajorFormatter = FormatStrFormatter('%.2f%%')
    
    fig = plt.figure(figsize=(12,8))
    ax = fig.add_subplot(1, 1, 1)
    ax.xaxis.set_major_locator(xmajorLocator)
    ax.xaxis.set_major_formatter( xmajorFormatter)
    ax.yaxis.set_major_formatter(ymajorFormatter) 

    for index in styleResult:
        exRet =((styleResult[index]['cumulative_returns'] - styleResult[index]['benchmark_cumulative_returns'])*100).tolist()
        plt.plot(dates,exRet, linewidth = linewidth)

    plt.grid(True)
    plt.legend(styleResult.keys(),loc='best',prop = legendFont)
    plt.tick_params(labelsize=0.9*labelsize)
    plt.title(u'风格因子累计超额收益曲线',fontsize = fontsize,verticalalignment = 'bottom',horizontalalignment = 'center',fontproperties = titleFont)
    plt.xlabel(u'时间日期',fontsize=fontsize,verticalalignment = 'top', horizontalalignment = 'center',fontproperties = scaleFont)
    plt.ylabel(u'超额收益百分比', fontsize = fontsize,  verticalalignment = 'bottom', horizontalalignment = 'center',rotation=90, fontproperties = scaleFont)
    pass

style_factor_performance()
````
