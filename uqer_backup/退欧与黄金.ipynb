### 1.黄金期货合约与现货交易
  可以看到，英国脱欧与国际货币形势的不稳定大幅推升了黄金期货与现货合约价格，2016年7月到10月交割的期货合约涨幅都超过3%，现货涨幅超过4%.
  但值得注意的是，今天也出现了两周以来首次现货价格超过期货价格的情况，现货价比7月、8月、9月交割的期货价都高，只比10月份交割期货价略低，基本表明期货市场对事件反应迟缓或现货市场反应过激，预期经历本周末的冷静期后，这一现象将被调整，期货与现货价差将回到合理区间。
  个人看法是现货市场不存在涨跌停制度，因而出现了一定程度的过激反应，在市场情绪稳定后相对期货的价格应该会有一定下调。

### 2.黄金有关概念股行情
  根据申万行业分类，黄金作为二级类别共有以下成员股票：中润资源、中金黄金、刚泰控股、商赢环球、园城黄金、山东黄金、恒邦股份、湖南黄金、紫金矿业、荣华实业、西部黄金、赤峰黄金。这些概念股除商赢环球停牌外，有6只涨停，3只涨幅超过6%，剩下两只涨幅2%到3%.
  这里值得注意的地方在于，A股仍然表现出了情绪化的特质，“园城黄金”早被借壳，其主营业务与黄金并无关联，而主要是房地产开发经营，在今天地产板块表现一般的情况下被推升8%，而刚泰控股这一从事矿业投资和贵金属投资的企业涨幅也只有3%. 这是未来市场趋于稳定时值得注意的情况，黄金概念过去后，园城黄金会不会守住今天的涨幅呢？
  另外需要关注的地方在于，其中的一些股票与黄金期货合约和现货价格走势相关性很高，若能发现某一市场行情出现在另一市场之前，将可能有机会构造相关策略进行套利。想法也不成熟，姑妄听之。

```python
goldSec=set_universe(IndSW.HuangJinL2)
DataAPI.EquGet(secID=goldSec,field=u"ticker,secShortName,secFullName,primeOperating")
```

```python
# 计算累计收益
df=DataAPI.MktEqudGet(secID=goldSec,beginDate=u"20160610",field="preClosePrice,closePrice,secShortName,ticker,tradeDate")
df = df[df.secShortName.isin(['西部黄金','中金黄金','山东黄金','赤峰黄金'])]
df["ret"]=df["closePrice"]/df["preClosePrice"]-1
histPrice=df.pivot(index="tradeDate",columns="secShortName",values="ret")

histPrice['tt中金黄金'] = np.cumsum(histPrice['中金黄金'])
histPrice['tt山东黄金'] = np.cumsum(histPrice['山东黄金'])
histPrice['tt西部黄金'] = np.cumsum(histPrice['西部黄金'])
histPrice['tt赤峰黄金'] = np.cumsum(histPrice['赤峰黄金'])
histPrice['山东-中金'] = histPrice['tt山东黄金'] - histPrice['tt中金黄金']
histPrice['西部-赤峰'] = histPrice['tt西部黄金'] - histPrice['tt赤峰黄金']

fig = plt.figure() # Create matplotlib figure
ax = fig.add_subplot(111) # Create matplotlib axes
fig2 = plt.figure() # Create matplotlib figure
ax2 = fig2.add_subplot(111)  #ax.twinx() # Create another axes that shares the same x-axis as ax.
fig3 = plt.figure() # Create matplotlib figure
ax3 = fig3.add_subplot(111)

histPrice.plot(y=['中金黄金','山东黄金','西部黄金','赤峰黄金'],ax=ax, figsize=(18, 4),  linewidth='2')
report.fig_style(ax, [u'中金黄金',u'山东黄金',u'西部黄金',u'赤峰黄金'],'upper right')

histPrice.plot(y=['tt中金黄金','tt山东黄金','tt西部黄金','tt赤峰黄金'],ax=ax2, figsize=(18, 4),  linewidth='2')
report.fig_style(ax2, [u'tt中金黄金',u'tt山东黄金',u'tt西部黄金',u'tt赤峰黄金'],'upper right')

histPrice.plot(y=['山东-中金','西部-赤峰'],ax=ax3, figsize=(18, 4),  linewidth='2')
report.fig_style(ax3, [u'山东-中金',u'西部-赤峰'],'upper right')

histPrice.tail(5)
```

```python
startDate = '20160601'
endDate = dt.date.today().strftime('%Y%m%d')
#现货黄金
sPrice=DataAPI.MktSpotdGet(secID=u"",ticker=u"Au(T%2bD)",beginDate=startDate, field=u"secShortName,closePrice,tradeDate,exchangeCD,ticker",pandas="1")
spotPrice=sPrice.pivot(index="tradeDate",columns="ticker",values="closePrice")
#期货黄金连续
mktTrend=DataAPI.MktMFutdGet(contractObject=u"AU",contractMark='L0',startDate=startDate, endDate=u"",field=u"secShortName,ticker,tradeDate,closePrice,settlePrice,turnoverValue",pandas="1")
fut=mktTrend.pivot_table(index="tradeDate",columns="ticker",values="closePrice",fill_value=0.0)
#黄金ETF,(博时黄金ETF, 沪深300ETF)
tmp=DataAPI.FundNavGet(beginDate=startDate,ticker=u"159937,159919",field=u"NAV,endDate,ticker,secShortName")
etfTrend=tmp.pivot(index="endDate",columns="secShortName",values="NAV")
#黄金股票
goldSec = ['600547'] # 山东黄金
df=DataAPI.MktEqudGet(ticker=goldSec,beginDate=startDate,field="ticker,tradeDate,secShortName,preClosePrice,closePrice")
#df["ret"]=df["closePrice"]/df["preClosePrice"]-1
stock=df.pivot(index="tradeDate",columns="secShortName",values="closePrice")

td = pd.DataFrame()
td['fut']  = fut.sum(axis=1)
td["spot"] = spotPrice
td['stock'] = stock
td['etf'] = etfTrend['博时黄金ETF']
td['hs300'] = etfTrend['嘉实沪深300ETF']

td['spread'] = 100.0 *( td.fut / td.spot -1.0)
td['fut_pct'] = 100* td.fut.pct_change()
td['spot_pct'] = 100* td.spot.pct_change()
td['stock_pct'] = 100* td.stock.pct_change()
td['etf_pct'] = 100* td.etf.pct_change()
td['hs300_pct'] = 100* td.hs300.pct_change()

td['fut_sum'] = td.fut_pct.cumsum()
td['spot_sum'] = td.spot_pct.cumsum()
td['stock_sum'] = td.stock_pct.cumsum()
td['etf_sum'] = td.etf_pct.cumsum()
td['hs300_sum'] = td.hs300_pct.cumsum()

fig = plt.figure() # Create matplotlib figure
ax = fig.add_subplot(111) # Create matplotlib axes
ax2 = ax.twinx()

fig3 = plt.figure() # Create matplotlib figure
ax3 = fig3.add_subplot(111) # Create matplotlib axes
ax4 = ax3.twinx()

td.plot(y=['fut_pct','spot_pct','etf_pct'],ax=ax, figsize=(18, 4),  linewidth='2',color='gbk' )
report.fig_style(ax, [u'期货',u'现货',u'基金'],'upper right')
td.plot(y=['stock_pct','hs300_pct'],ax=ax2, figsize=(18, 4),  linewidth='2' , color='ry')
report.fig_style(ax2, [u'股票',u'指数'],'upper left')

td.plot(y=['fut_sum','spot_sum','etf_sum'],ax=ax3, figsize=(18, 4),  linewidth='2',color='gbk')
report.fig_style(ax3, [u'期货',u'现货',u'基金'],'upper left')
td.plot(y=['stock_sum','hs300_sum'],ax=ax4, figsize=(18, 4),  linewidth='2' , color='ry')
report.fig_style(ax4, [u'股票',u'指数'],'upper right')

td.tail(5)
```
