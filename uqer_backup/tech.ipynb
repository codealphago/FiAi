[`用量化公式追逐巴菲特选股的阿尔法之旅(四因素模型)`]
基于规模、价值、动量以及质量因子的投资策略，他们把这些介于传统的CAPM贝塔和阿尔法之间的新策略统称为“聪明贝塔”策略
1. 范围(行业主题宏观):(规模,板块[创业板,主板,中小版],行业,主题[环保,新能源,工业4,机器人AI])   
2. 风格(因素分析):(价值,成长,周期，性质[央企,国企,私企,外企])
3. 质量(基本面):(高利润,高成长,稳定和优良的管理,股息收益率,自由现金流收益率),
4. 动量(技术面):(波动率,技术指标[价格逐步上涨,MACD,RSI])

[`策略组合选择`]
- 规模投资: [`买入小盘股`],      发达市场小盘股正阿尔法在学术成果公开后基本消失了, 可以得和风险匹配的回报. 
- 价值投资: [`买入低估值的股票`],  低估值的公司未必是高质量的公司,价格可能长期低迷,落入价值陷阱. 
- 动量投资: [`买入过去价格上涨的股票`], 回报远高于价值投资,但需要判定市场方向,执行难度大 
- 质量投资: [`买入高质量股票`],       质量体现在高利润,高成长,稳定和优良的管理,股票价格偏高,但有正的阿尔法.
- 规模+价值: [`买入小盘低估股票`],    长期表现高于大盘高估的股票.
- 价值+动量: [`买入因外部事件而价格逐步上涨的低估股票`],   互补关系可以享受双重好处
- 价值+质量: [`买入具有合理价格的高质量股票`],     单纯价值投资可能错过这些估值相对较高的股票.
- 价值+质量+动量: [`买入价格逐步上涨,但其估值仍然低于高质量应得价格的股票`],  互补关系,享受三重好处

```python

#判断是否多头趋势
def duotou_5_10(seclist,trade_date = dt.date.today()):
    s_date = comm.getPrevBusinessDay(trade_date,1).strftime("%Y%m%d")          #计算出倒数第一个交易日
    duotou = []
    if seclist is None or len(seclist) == 0:
        return duotou
    kLine = DataAPI.MktStockFactorsOneDayGet(tradeDate=s_date,ticker=seclist,field=['ticker','secID','MA5','MA10'],pandas="1")
    kLine = kLine.dropna()
    for stock, ma5, ma10 in zip(kLine['secID'].values, kLine['MA5'].values, kLine['MA10'].values):
        if ma5 > ma10:
            duotou.append(stock)
    return duotou
# doutou =  duotou_5_10(stockList)
# print doutou

#判断是否为底部形态，判断标准为股价相对于近期(20交易日)高点大幅下跌超过downPercent，并且收盘价在五日线十日线之下，并且收红或者收星，跌幅小于7%

def isButtom(seclist,trade_date = dt.date.today(), downPercent=0.3):
    s_date = comm.getPrevBusinessDay(trade_date,1).strftime("%Y%m%d")          #计算出倒数第一个交易日
    daysAhead = comm.getPrevBusinessDay(trade_date,20).strftime("%Y%m%d")          #计算出倒数第一个交易日
    rs = []
    dayInfo = DataAPI.MktEqudAdjGet(ticker=seclist, beginDate=daysAhead, endDate=s_date ,field=['ticker','secID', 'openPrice', 'closePrice', 'preClosePrice'],pandas="1")
    dayInfo.dropna()
    for stock in seclist:
        stockDayInfo = dayInfo[dayInfo['ticker']==stock]
        closePrices = stockDayInfo['closePrice'].values
        ma5 = np.mean(closePrices[-5:])
        ma10 = np.mean(closePrices[-10:])
        maxClosePrice = np.max(closePrices)        
        curr_price = closePrices[-1]
        openPrice = stockDayInfo['openPrice'].values[-1]        
        preClosePrice = stockDayInfo['preClosePrice'].values[-1]
        if ( maxClosePrice-curr_price)  /maxClosePrice > downPercent and curr_price < ma5 and ma5 < ma10 and (curr_price > openPrice or abs(curr_price-openPrice)/openPrice < 0.02) and abs(curr_price-preClosePrice)/preClosePrice<0.07 :
            rs.append(stock)
    return rs
#bottom =  isButtom(stockList)
#print bottom
```

```python
'''
Kelly Formula在仓位控制中的应用
在股票市场中，kelly formula的形式和赌博中稍有不同，因为失败之后有一个expected loss,  我们将公式修改为 f = (bp - cq) / (bc)，来做一个简单的回测。
假定赚钱(p)赔钱概率(q = 1-p)一样，均为0.5，我们认为净赢率(b)为最高价/当前价-1，净损失率(c)为1-最低价/当前价。
如果当前价大于一年内最高价，满仓，如果当前价小于一年内最低价，空仓，其他情况使用之前得到的f计算出持仓比例(0-1之间)，每周调仓。

'''
# #计算凯利公式
# f* 为现有资金应进行下次投注的比例: f = (bp - cq) / (bc)
# b 为净盈率:   最高价/当前价-1
# c 为净损失率:  1-最低价/当前价
# p 为获胜率 :   
# q 为落败率，即 1 - p； 
def kelly_f(b,c,p):
    q = 1- p
    f = round((b*p-c*q)/ (b*c),2)
    return f

def kelly_d(ticker,trade_date = dt.date.today() ,back_window=250):
    #获取量价数据
    s_date = comm.getPrevBusinessDay(trade_date,back_window).strftime("%Y%m%d")   
    e_date = comm.getPrevBusinessDay(trade_date).strftime("%Y%m%d")          
    dayInfo = DataAPI.MktEqudAdjGet(ticker=ticker, beginDate=s_date, endDate=e_date ,field=['secID','tradeDate','secShortName', 'closePrice', 'preClosePrice'],pandas="1")
    dayInfo.dropna()
    #print dayInfo.tail(10)
    #历史数据统计过去一年中日内的涨跌概率
    wp =  1.00 - len(dayInfo[dayInfo.closePrice -  dayInfo.preClosePrice >=0 ])*1.00/len(dayInfo)
    #取窗口内最高价| 最低价
    HH = np.max(dayInfo.closePrice.values)
    LL = np.min(dayInfo.closePrice.values)
    CC = dayInfo.closePrice.values[-1]
    B = (HH/CC) -1
    C =1- (LL/CC)
    #计算凯利公式
    F= kelly_f(b=B,c=C,p=wp)
    #限定F的范围
    if (F < 0.0):
        F = 0.0
    if (F > 1.0):
        F = 1.0
        
    if (CC >= HH):
        F = 1.0
    if (CC <= LL):
        F = 0.0
   
    print "HH = %f, LL = %f, CC = %f, B=%f, C=%f, p=%f, F = %f"%(HH, LL, CC, B,C,wp, F)
    return F

# 改进#p，q并不都取0.5，而是根据历史数据统计过去一年中5个交易日内的涨跌概率。
# 加入新的标准，统计过去一年内该股票的一周最大涨跌幅，以此计算b和c，得到的f与之前f的求平均。
# ???设定止损。如果一周内跌幅超过8%，则清仓。
def kelly_w(ticker,trade_date = dt.date.today(),back_window=250,weight=0.5):
    #获取量价数据
    s_date = comm.getPrevBusinessDay(trade_date,back_window).strftime("%Y%m%d")   
    e_date = comm.getPrevBusinessDay(trade_date).strftime("%Y%m%d")          
    dayInfo = DataAPI.MktEqudAdjGet(ticker=ticker, beginDate=s_date, endDate=e_date ,field=['secID','tradeDate','secShortName', 'closePrice', 'preClosePrice'],pandas="1")
    dayInfo.dropna()
    print dayInfo
    prices = dayInfo.closePrice.values
    print prices
    max_rise ,max_drop,win_times,lose_times = 0,1,0,0
    for i in range(len(prices) -4 ):
        weekly = prices[i:i+4]
        max5 = np.max(weekly)
        min5 = np.min(weekly)
        temp_max_rise = np.log(max5/weekly[0])
        temp_max_drop = np.log(min5/weekly[0])
        if max_rise < temp_max_rise:
            max_rise = temp_max_rise
        if max_drop > temp_max_drop:
            max_drop = temp_max_drop
        if weekly[0]*1.005 < weekly[-1]:
            win_times = win_times + 1
        if weekly[0]*1.005 > weekly[-1]:
            lose_times = lose_times + 1
    #历史数据统计过去一年中5个交易日内的涨跌概率
    wp =  1.00*win_times/(win_times + lose_times)
    #计算kelly old
    HH = np.max(dayInfo.closePrice.values)
    LL = np.min(dayInfo.closePrice.values)
    CC = dayInfo.closePrice.values[-1]
    B1 = (HH/CC) -1
    C1 = 1- (LL/CC)
    FD= kelly_f(b=B1,c=C1,p=wp)
    if (FD < 0.0):
        FD = 0.0
    if (FD > 1.0):
        FD = 1.0
    print "B1=%f, C1=%f, p=%f, FD = %f"%(B1,C1,wp, FD)
    
    #取窗口内最高价| 最低价
    B2 = max_rise
    C2 = -max_drop
   
    FW= kelly_f(b=B2,c=C2,p=wp)
    print "B2=%f, C2=%f, p=%f, FW = %f"%( B2,C2,wp, FW)

    #限定FW的范围
    if (FW < 0.0):
        FW = 0.0
    if (FW > 1.0):
        FW = 1.0
        
    F = round(FW * weight + FD * (1-weight),2)   
    if (CC >= HH):
        F = 1.0
    if (CC <= LL):
        F = 0.0
    #print "date: ", e_date
    return F
# print kelly_f(0.5,0.4,0.9)
# kelly_d('600030',back_window = 20)
# kelly_w('300027',dt.datetime.strptime('20141205','%Y%m%d'), back_window = 20)
```

```python
#波动率
#过去上涨好的股票（趋势）
#MACD法则
def selectMACD(seclist,trade_date = dt.date.today()):
    hist_days = 120
    # MACD计算参数
    short_win = 12    # 短期EMA平滑天数
    long_win  = 26    # 长期EMA平滑天数
    macd_win  = 9     # DEA线平滑天数
    long_bucket = []  
    short_bucket = []
    s_date = comm.getPrevBusinessDay(trade_date,hist_days).strftime("%Y%m%d")          #计算出倒数第一个交易日
    e_date = comm.getPrevBusinessDay(trade_date,1).strftime("%Y%m%d")     
    for stk in seclist:
        # 使用TA-Lib，从收盘价计算macd
        h_price = DataAPI.MktEqudGet(secID=stk,beginDate=s_date , endDate= e_date ,field="secID,closePrice")
        dif,dea,macd = ta.MACD(h_price.closePrice.values, fastperiod=short_win, slowperiod=long_win, signalperiod=macd_win)  
        # DIF、DEA均为正，DIF向上突破DEA，发出买入信号
        if dif[-1] > 0 and dea[-1] > 0 and macd[-2] < 0 and macd[-1] > 0:
            long_bucket.append(stk)
        # DIF、DEA均为负，DIF向下突破DEA，发出卖出信号
        if dif[-1] < 0 and dea[-1] < 0 and macd[-2] > 0 and macd[-1] < 0:
            short_bucket.append(stk)   
    return long_bucket, short_bucket
# long_list, short_list = selectMACD(stockList)
# print long_list
# print short_list
```

## 【成交量热度指标】

- 将标的上一个交易日的成交量在过去一段历史时间窗口的排名作为成交量热度指标，若排名靠前则认为标的在当前市场热度高，正处于短期放量的“炒作”时期，是正向热度；若排名靠后则认为该标的当前市场热度低，处于“萎靡”时期，是反向热度。

- 我们对正向热度和反向热度的信号分别作择时检验，并研究其所给能出的信息量。

```python
def VolHot(secid, type = 'S',  tdate = dt.date.today(), histLength = 20, dataWindow = 30): 
    sdate = comm.getPrevBusinessDay(tdate,histLength+dataWindow).strftime("%Y%m%d") 
    edate = comm.getPrevBusinessDay(tdate).strftime("%Y%m%d") 
    if type == 'I':
        origData = DataAPI.MktIdxdGet(ticker = secid,beginDate = sdate,endDate = edate, field = 'ticker,tradeDate,turnoverVol,openIndex,closeIndex,preCloseIndex',pandas = '1')
    else :
        origData = DataAPI.MktEqudAdjGet(ticker = secid,beginDate = sdate,endDate = edate, field = 'ticker,tradeDate,turnoverVol,openPrice,closePrice,preClosePrice',pandas = '1')
    signal = []
    for index in origData['tradeDate']:
        tmpData = origData[origData.tradeDate <= index][-histLength:]
        if len(tmpData)< histLength:
            signal.append(0)
        else:
            tmpData = tmpData['turnoverVol'].tolist()
            rank = sorted(cp.deepcopy(tmpData),reverse = True).index(tmpData[-1])

            if rank < histLength/5:
                signal.append(1)
            elif rank > 4.0*histLength/5:
                signal.append(-1)
            else:
                signal.append(0)
    origData['signal'] = signal
    return origData[20:]
#VolHot('000009',dataWindow = 30)
```
