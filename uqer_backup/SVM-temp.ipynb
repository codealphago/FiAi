```python
def train_data(universe,used_factors,start = '2014-05-01',end = '2014-10-30',ad_period=10):
    import pandas as pd
    import numpy as np
    from datetime import datetime
    from datetime import timedelta
    from CAL.PyCAL import *

    cal = Calendar('China.SSE')
    date=cal.bizDatesList(start,end)
    
    set_factors=['tradeDate','secID']
    
    training_data_set=pd.DataFrame()
    for i,idate in enumerate(date):
        if i+ad_period>=len(date):
            continue
        t=i+ad_period
        tdate=date[t]
        tmp_price=DataAPI.MktEqudGet(tradeDate=idate,secID=universe,isOpen="",field=u"secID,tradeDate,closePrice",pandas="1")
        tmp_t_price=DataAPI.MktEqudGet(tradeDate=tdate,secID=universe,isOpen="",field=u"secID,tradeDate,closePrice",pandas="1")
        tmp_price['updown']=(tmp_t_price['closePrice']-tmp_price['closePrice'])/tmp_price['closePrice']
        #print tmp_price
        for i in tmp_price.index:
            if tmp_price.ix[i,['updown']].values>=0.08:
                tmp_price.ix[i,['updown']]=2
            elif tmp_price.ix[i,['updown']].values>=0.03 and tmp_price.ix[i,['updown']].values<=0.08:
                tmp_price.ix[i,['updown']]=1
            elif tmp_price.ix[i,['updown']].values<=0.03 and tmp_price.ix[i,['updown']].values>=-0.03:
                tmp_price.ix[i,['updown']]=0
            elif tmp_price.ix[i,['updown']].values<=-0.03 and tmp_price.ix[i,['updown']].values>=-0.08:
                tmp_price.ix[i,['updown']]=-1
            elif tmp_price.ix[i,['updown']].values<=-0.08:
                tmp_price.ix[i,['updown']]=-2
        tmp_price['updown']=tmp_price['updown'].astype(int)
        tmp=DataAPI.MktStockFactorsOneDayGet(tradeDate=idate,secID=universe,field=set_factors+used_factors,pandas="1")
        tmp=pd.merge(tmp,tmp_price)
        tmp=tmp.set_index(['secID'])
        tmp=tmp.dropna()
        for ifac in used_factors:
            tmp[ifac]=pd.Series(standardize(neutralize(winsorize(tmp[ifac].to_dict(),n_draw=5),idate.strftime('%Y%m%d'))))  
        training_data_set=pd.concat([training_data_set,tmp])
    rownum=range(len(training_data_set))
    rownum=pd.Series(rownum)
    training_data_set=training_data_set.set_index(rownum)  
    #print training_data_set.head()

    Xtrain=training_data_set.loc[:,used_factors]
    Xtrain=np.mat(Xtrain)
    Ytrain=training_data_set.loc[:,'updown']
    Ytrain=np.mat(Ytrain).T
    #print 'Data acquisition has been completed .'
    return Xtrain,Ytrain    
```

```python
from CAL.PyCAL import *
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
from datetime import datetime
from pandas.tseries.offsets import Day
starttime = datetime.now()

cal = Calendar('China.SSE')
start='2014-01-01'
#end='2015-12-31'
end='2014-04-30'
date=cal.bizDatesList(start,end)

universe=set_universe('SH50')

#used_factors=['CCI5','CCI88','SRMI','CMO','VEMA10','VMACD','ASI','Volatility']
used_factors = ['MA10', 'MA60', 'ROA', 'PE', 'LCAP', 'DHILO', 'DebtEquityRatio', 'OperatingProfitGrowRate', 'TotalAssetGrowRate', 'NPToTOR']
set_factors=['tradeDate','secID']
price_factors=['']

training_data_set=pd.DataFrame()

#偏移时间
ad_period=10

for i,idate in enumerate(date):
    if i+ad_period>=len(date):
        continue
    
    t=i+ad_period
    tdate=date[t]
    tmp_price=DataAPI.MktEqudGet(tradeDate=idate,secID=universe,isOpen="",field=u"secID,tradeDate,closePrice",pandas="1")
    tmp_t_price=DataAPI.MktEqudGet(tradeDate=tdate,secID=universe,isOpen="",field=u"secID,tradeDate,closePrice",pandas="1")
    tmp_price['updown']=(tmp_t_price['closePrice']-tmp_price['closePrice'])/tmp_price['closePrice']
    for i in tmp_price.index:
        if tmp_price.ix[i,['updown']].values>=0.08:
            tmp_price.ix[i,['updown']]=2
        elif tmp_price.ix[i,['updown']].values>=0.03 and tmp_price.ix[i,['updown']].values<=0.08:
            tmp_price.ix[i,['updown']]=1
        elif tmp_price.ix[i,['updown']].values<=0.03 and tmp_price.ix[i,['updown']].values>=-0.03:
            tmp_price.ix[i,['updown']]=0
        elif tmp_price.ix[i,['updown']].values<=-0.03 and tmp_price.ix[i,['updown']].values>=-0.08:
            tmp_price.ix[i,['updown']]=-1
        elif tmp_price.ix[i,['updown']].values<=-0.08:
            tmp_price.ix[i,['updown']]=-2
    tmp_price['updown']=tmp_price['updown'].astype(int)
    
    tmp=DataAPI.MktStockFactorsOneDayGet(tradeDate=idate,secID=universe,field=set_factors+used_factors,pandas="1")
    tmp=pd.merge(tmp,tmp_price)
    tmp=tmp.set_index(['secID'])
    tmp=tmp.dropna()

    for ifac in used_factors:
        tmp[ifac]=pd.Series(standardize(neutralize(winsorize(tmp[ifac].to_dict(),n_draw=5),idate.strftime('%Y%m%d'))))  
    training_data_set=pd.concat([training_data_set,tmp])
    
rownum=range(len(training_data_set))
rownum=pd.Series(rownum)
training_data_set=training_data_set.set_index(rownum)  

Xtrain=training_data_set.loc[:,used_factors]
Xtrain=np.mat(Xtrain)
Ytrain=training_data_set.loc[:,'updown']
Ytrain=np.mat(Ytrain).T
print 'Data acquisition has been completed .'
```
