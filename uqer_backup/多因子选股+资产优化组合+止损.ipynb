```python
from pandas import DataFrame
import numpy as np
import pandas as pd
import scipy as sp
import scipy.optimize as sco
from datetime import *

def get_stockset(universe,sdate,edate):
    
    seclist = []
    for s in universe: seclist.append(s[0:6])
    listStock = DataAPI.SecIDGet(ticker=seclist,assetClass='E',field=u"secID,listDate",pandas="1")
    listStock = listStock[listStock.listDate < sdate.strftime('%Y-%m-%d') ]
    
    p=DataAPI.MktStockFactorsOneDayGet(tradeDate=edate,secID=listStock.secID.values,field=u"secID,LFLO,FixAssetRatio,pe,EquityToAsset,NetProfitRatio",pandas="1")
    effective_factors={'LFLO':False,'FixAssetRatio':False,'PE':False,'EquityToAsset':True,'NetProfitRatio':True}
    p.index=p['secID']
    score={}
    for fac,value in effective_factors.items():
        score[fac]=p[fac].rank(ascending=value,method='first')
    buy_stock_set = list((DataFrame(score)*np.array([3,4,2,1,5])).T.sum().order(ascending = False).head(40).index)
    return buy_stock_set[:8]

edate = datetime.strptime('2015-06-10','%Y-%m-%d')
dayspan=timedelta(days=300)
sdate = edate - dayspan

universe = set_universe('HS300')  + set_universe('ZZ500') + set_universe('CYB') 
ticker = get_stockset(universe,sdate,edate)
print ticker
```

```python
def get_historical_closes(ticker, start_date, end_date):
    p = DataAPI.MktEqudGet(secID=ticker,beginDate=start_date,endDate=end_date,field=u"secID,tradeDate,closePrice",pandas="1") 
    p.rename(columns={'tradeDate': 'Date', 'closePrice': 'Close'}, inplace=True)
    pivoted = p.pivot(index='Date', columns='secID')
    pivoted.columns = pivoted.columns.droplevel(0)
    return np.log(pivoted/pivoted.shift(1))

# 2014-08-14 00:00:00 2015-06-10 00:00:00
# ['600382.XSHG', '002113.XSHE', '300223.XSHE', '002762.XSHE', '600883.XSHG', '300475.XSHE', '600381.XSHG', '600890.XSHG']
# 1202
# ticker = ['600382.XSHG', '002113.XSHE', '300223.XSHE', '002762.XSHE', '600883.XSHG', '300475.XSHE', '600381.XSHG', '600890.XSHG']

get_historical_closes(ticker,'2014-08-14','2015-06-10')

```
