[`用量化公式追逐巴菲特选股的阿尔法之旅(四因素模型)`]
基于规模、价值、动量以及质量因子的投资策略，他们把这些介于传统的CAPM贝塔和阿尔法之间的新策略统称为“聪明贝塔”策略
1. 范围:(规模,板块[创业板,主板,中小版],行业,主题[环保,新能源,工业4,机器人AI])   
2. 风格:(价值,成长,周期，性质[央企,国企,私企,外企])
3. 质量:(高利润,高成长,稳定和优良的管理,股息收益率,自由现金流收益率)
4. 动量:(波动率,技术指标[价格逐步上涨,MACD,RSI])

[`策略组合选择`]
- 规模投资: [`买入小盘股`],      发达市场小盘股正阿尔法在学术成果公开后基本消失了, 可以得和风险匹配的回报. 
- 价值投资: [`买入低估值的股票`],  低估值的公司未必是高质量的公司,价格可能长期低迷,落入价值陷阱. 
- 动量投资: [`买入过去价格上涨的股票`], 回报远高于价值投资,但需要判定市场方向,执行难度大 
- 质量投资: [`买入高质量股票`],       质量体现在高利润,高成长,稳定和优良的管理,股票价格偏高,但有正的阿尔法.
- 规模+价值: [`买入小盘低估股票`],    长期表现高于大盘高估的股票.
- 价值+动量: [`买入因外部事件而价格逐步上涨的低估股票`],   互补关系可以享受双重好处
- 价值+质量: [`买入具有合理价格的高质量股票`],     单纯价值投资可能错过这些估值相对较高的股票.
- 价值+质量+动量: [`买入价格逐步上涨,但其估值仍然低于高质量应得价格的股票`],  互补关系,享受三重好处

```python


#基本筛选（排除ST，流动性差的,新上市,刚复牌,弱势股票）
def removeFilter(stocklist,trade_date=dt.date.today()):
    trade_date = comm.getPrevBusinessDay(trade_date,0)
    # 去除ST股
    #print len(stocklist)
    stocklist = filt.removeST(stocklist,trade_date)
    #print len(stocklist)
    #去掉20个交易日之前的20个交易日平均日交易量小于1000万人民币的股票
    stocklist = filt.removeLiqid(stocklist,trade_date)
    #print len(stocklist)
    # 去除新上市或复牌的股票
    stocklist = filt.removeNew(stocklist,trade_date)
    #print len(stocklist)
    # 去除弱势股票
    stocklist = filt.removeWeak(stocklist,trade_date)
    #print len(stocklist)
    return stocklist

#stks = removeFilter(stockList)
#print stks

#流通市值(总市值)较小的股票(前length=50只)
def marketValueFilter(stocklist,trade_date=dt.date.today(),length=50):
    last_day = comm.getPrevBusinessDay(trade_date,1)          
    last_day_str = last_day.strftime("%Y%m%d")
    mkt_value = DataAPI.MktEqudGet(secID=stocklist,tradeDate=last_day_str,field="secID,secShortName,negMarketValue,marketValue")
    mkt = mkt_value.sort('marketValue')
    #print mkt
    if (length > len(mkt) ):
        length = len(mkt)
    stocklist = mkt.secID[0:length]
    return stocklist

#print marketValueFilter(stockList)

#随机选取length 50只，作为业绩比较)
def randomFilter(stocklist,length=10):
    if (length > len(stocklist) ):
        length = len(stocklist)
    stocklist = rd.sample(stocklist,length)
    return stocklist
#randomFilter(stockList)
    
```

```python
#算A股中位数作为参考值  LCAP：对数市值，DebtEquityRatio:产权比率，ROE:权益回报率，NetProfitGrowRate：净利润增长率，ROA：资产回报率     
def getMarketMedian(benchmark , trade_date =dt.date.today()):
    trade_date = comm.getPrevBusinessDay(trade_date,1)
    sdate = trade_date.strftime("%Y%m%d")
    data = DataAPI.MktStockFactorsOneDayGet(tradeDate = sdate,secID=benchmark,field=u"secID,LCAP,DebtEquityRatio,ROE,ROE5,NetProfitGrowRate,ROA,ROA5,PE,PB,PCF,PS",pandas="1")
    data.set_index('secID',inplace=True)
    ans = data.median() 
    return ans
#print getMarketMedian(stockList)

#估值（价值）筛选-----------------------------------------------------
def jiazhiFilter(seclist,trade_date =dt.date.today(),max_out=50 ):
    jiazhi_list = []
    ans = getMarketMedian(seclist,trade_date)
    trade_date = comm.getPrevBusinessDay(trade_date,1)
    sdate = trade_date.strftime("%Y%m%d")
    data = DataAPI.MktStockFactorsOneDayGet(tradeDate = sdate,secID=seclist,field=u"secID,PE,PB,PCF,PS",pandas="1")
    data.set_index('secID',inplace=True)
    #print data
    # 条件1  PE小于市场平均值
    # 条件2  PB小于市场平均值
    # 条件3  PCF小于市场平均值
    # 条件4  PS小于市场平均值
    for s in seclist : 
        if (  data['PE'][s] > 0 and data['PE'][s] < ans['PE'] and
            data['PE'][s] > 0 and data['PB'][s] < ans['PB'] and
            data['PCF'][s] > 0 and data['PCF'][s] < ans['PCF'] and
            data['PS'][s] > 0 and data['PS'][s] < ans['PS'] ):
            jiazhi_list.append(s)
    return jiazhi_list 
#jiazhi = jiazhiFilter(stockList)
#print len(jiazhi)  

#报表（成长）筛选------------------------------------------
def reportFilter(seclist,trade_date =dt.date.today() ):
    report_list = []
    ans = getMarketMedian(seclist,trade_date)
    trade_date = comm.getPrevBusinessDay(trade_date,1)
    sdate = trade_date.strftime("%Y%m%d")
    data = DataAPI.MktStockFactorsOneDayGet(tradeDate = sdate,secID=seclist,field=u"secID,LCAP,DebtEquityRatio,ROE,ROE5,NetProfitGrowRate,ROA,ROA5",pandas="1")
    data.set_index('secID',inplace=True)
    #print data
    # 条件1  总市值大于市场平均值？？？？
    # 条件2  产权比率小于市场平均值
    # 条件5  ROE大于市场平均值
    # 条件6  净利润增长率大于市场平均值
    # 条件7  资产回报率大于市场平均值
    for s in seclist : 
        if (  data['LCAP'][s] <= ans['LCAP'] and
            #data['DebtEquityRatio'][s] < ans['DebtEquityRatio'] and
            data['ROE5'][s] > 0 and data['ROE5'][s]> ans['ROE5'] and
            data['ROE'][s] > 0 and data['ROE'][s]> 1.5 *ans['ROE'] and
            data['NetProfitGrowRate'][s] > 0 and data['NetProfitGrowRate'][s] > 2 * ans['NetProfitGrowRate'] and
            data['ROA5'][s] > 0 and data['ROA5'][s] >=  ans['ROA5'] and
            data['ROA'][s] > 0 and data['ROA'][s] >= 1.5 * ans['ROA'] ):
            report_list.append(s)
    return report_list 
#report_list = reportFilter(stockList)
#print len(report_list)

#每个季度，计算沪深300成分股资产负债表中的 现金及现金等价物/总资产 ，数值以最近一次披露的财报为准
#清仓，选出该比率最大的前30只股票，将资金分成30份，分别买入
#每60个交易日调仓一次
def getCashPct(seclist, date=dt.date.today(),max_out=50):
    start, end = (date - dt.timedelta(weeks=26)).strftime('%Y%m%d'), date.strftime('%Y%m%d')
    df = DataAPI.FdmtBSGet(secID=seclist, publishDateBegin=start, publishDateEnd=end, field=['secID', 'cashCEquiv', 'TAssets'])
    CashPct = {}
    for stock in seclist:
        try:
            df_sub = df[df.secID==stock]
            df_sub['pct'] = df_sub['cashCEquiv'] / df_sub['TAssets']
            CashPct[stock] = df_sub['pct'].mean()
        except:
            pass
    ret = pd.DataFrame(CashPct.items(),columns=['secID','cashPct'],index=CashPct.keys())
    ret = ret.sort('cashPct',ascending=False)
    if max_out < len(ret):
        seclist = ret.secID.values[0:max_out]
    else:
        seclist = ret.secID.values
    return seclist
#print getCashPct(stockList)
```

###`仓位管理`

```python
#卖出策略(不在买入的里面的pos就卖掉)
def sellStrategy(pos, buys):
    sells = []
    for s in pos:
        if s not in buys:
            sells.append(s)
    return sells
#print sellStrategy(['a','b'],['b','c','d'])

#买入策略(买入在pos中没有的)
def buyStrategy(pos, buys):
    ret = []
    for s in buys:
        if s not in pos:
            ret.append(s)
    return ret
#print buyStrategy(['a','b'],['b','c','d'])
```
