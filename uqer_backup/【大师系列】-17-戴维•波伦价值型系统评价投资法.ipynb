## 策略说明及数据来源
###### A、总市值大于市场平均值                 数据来源：MktStockFactorsOneDayGet，LCAP
###### B、产权比率小于市场平均值                数据来源：MktStockFactorsOneDayGet，DebtEquityRatio
###### C、每股企业自由现金流量大于市场平均值        数据来源：FdmtCFGet，报表计算
###### D、股价/每股自由现金流量小于市场平均值       数据来源：FdmtCFGet，报表计算   或者直接PCF
###### E、最近三年ROE大于市场平均值              数据来源：MktStockFactorsOneDayGet，ROE
###### F、最近三年净利润同比增长率大于市场平均值     数据来源：MktStockFactorsOneDayGet，NetProfitGrowRate
###### G、最近一年市盈率小于最近3年平均市盈率       数据来源：MktStockFactorsOneDayGet ，PE  选中股数太少本条略去
###### H、投入资本回报率大于市场平均值            数据来源：MktStockFactorsOneDayGet，ROA（用于替换付费的ROIC）
###### I、投入资本回报率大于市场基准利率           因ROIC值问题本条略去

除C、D外均可直接用MktStockFactorsOneDayGet中的相应因子。

```python
#每股现金流量是一个TTM值，计算公式是营业业务所带来的净现金流量/总股数，所以股价现金流量比即为总市值/营业业务所带来的净现金流量
def ncfOperateA(account):
    today = account.current_date
    cal = Calendar('China.SSE')
    lastDay = cal.advanceDate(today,'-1B',BizDayConvention.Unadjusted)
    today_str = lastDay.strftime("%Y%m%d")
    start_date = cal.advanceDate(lastDay,'-500B',BizDayConvention.Unadjusted)
    start_date_str = start_date.strftime("%Y%m%d")
    
    equity = pd.DataFrame()
    for rpt_type in ['Q1','S1','A']:
        try:
            tmp = DataAPI.FdmtCFGet(secID=account.universe, field=['secID', 'endDate', 'publishDate', 'NCFOperateA'], beginDate=start_date_str, publishDateEnd=today_str,  reportType=rpt_type)
        
        except:
            tmp = DataFrame()
        equity = pd.concat([equity, tmp], axis=0)

    equity = equity.sort(columns=['secID','endDate','NCFOperateA'], ascending=[True, False, False])
    equity = equity.drop_duplicates(cols=['secID','endDate'])
    equity = equity.dropna()
    
    signalNCFO = {}
    for stock in account.universe:
        try:
            num = len(equity[equity['secID']==stock]['endDate'])
            if (num>=4):
                endDate = equity[equity['secID']==stock]['endDate'].iloc[0]
                month= int(endDate[6])
                if (month==2) :
                    NCFO_TTM = equity[equity['secID']==stock]['NCFOperateA'].iloc[0]
                elif  (month==6):
                    NCFO_TTM = equity[equity['secID']==stock]['NCFOperateA'].iloc[0]-equity[equity['secID']==stock]['NCFOperateA'].iloc[3]+equity[equity['secID']==stock]['NCFOperateA'].iloc[2] 
                elif (month==3):
                    NCFO_TTM = equity[equity['secID']==stock]['NCFOperateA'].iloc[0]-equity[equity['secID']==stock]['NCFOperateA'].iloc[3]+equity[equity['secID']==stock]['NCFOperateA'].iloc[1]
                else:
                    NCFO_TTM = 'NA'
                if len(signalNCFO) == 0:
                    signalNCFO['secID'] = [stock]
                    signalNCFO['signalNCFO'] = [NCFO_TTM]
                else:
                    signalNCFO['secID'].append(stock)
                    signalNCFO['signalNCFO'].append(NCFO_TTM)    
        except:
            pass
 
    signalNCFO = DataFrame(signalNCFO)
    filterList1 = signalNCFO['secID'].tolist()
    
# 条件C：每股现金流量大于市场平均值

    signalNCFO =  signalNCFO[signalNCFO['signalNCFO'] > signalNCFO.median()['signalNCFO']]
    
    filter1_NCFO = {}
    for stock in filterList1:
        try:
            tmp = DataAPI.MktEqudGet(secID=stock,tradeDate=today_str,field='marketValue',pandas="1")
            price_NCFO = tmp['marketValue'].values[0]/signalNCFO[signalNCFO['secID'] == stock]['signalNCFO'].values[0]

            if len(filter1_NCFO) == 0:
                filter1_NCFO['secID'] = [stock]
                filter1_NCFO['price_NCFO'] = [price_NCFO]
            else:
                filter1_NCFO['secID'].append(stock)
                filter1_NCFO['price_NCFO'].append(price_NCFO)    
        except:
            pass

    filter1_NCFO = DataFrame(filter1_NCFO)
    
    filter1_NCFO = filter1_NCFO[filter1_NCFO['price_NCFO'] < np.mean(filter1_NCFO['price_NCFO'])]
    CFPSlist = filter1_NCFO['secID'].tolist()
    return CFPSlist
```
