```python
#股票价值计算---------------------
class StockValue:
    pcode = ''
    tdate = dt.date.today()
    sdate = '20110101'
    edate = ''
    his_price = pd.DataFrame()
    factor = pd.DataFrame()

    def __init__(self,code,trade_date = dt.date.today(), s_date = '20100101' ): 
        self.pcode = code
        self.tdate = trade_date
        self.sdate = s_date
        self.edate = comm.getPrevBusinessDay(self.tdate).strftime("%Y%m%d") 
        self.loaddata()
        
    def loaddata(self):
        #个股历史价格(不复权)
        self.his_price = DataAPI.MktEqudGet(ticker=self.pcode,tradeDate=self.edate,field=u"",pandas="1")
        self.his_price.index = self.his_price.tradeDate
        #print 'load price...'
        #个股因子数据
        self.factor = DataAPI.MktStockFactorsDateRangeGet(ticker=self.pcode,beginDate=self.sdate,endDate=self.edate,
                 field=['secID','tradeDate','PE','PB','EPS','EGRO','OperatingProfitRatio','OperatingProfitGrowRate','TotalProfitCostRatio','NPToTOR','ROE','ROA'],pandas="1")
        #print 'load factor...'
        self.factor.index = self.factor.tradeDate
       
    def displyData(self):  
        self.factor[['PE']].plot(figsize=(15, 4))
        self.factor[['NPToTOR','TotalProfitCostRatio','OperatingProfitRatio']].plot(figsize=(15, 4),title=u'net profit / total operating revenue  && total profit /(operating cost+finacal cost +selling cost  + management cost)')
        self.factor[['OperatingProfitGrowRate']].plot(figsize=(15, 4),title=U"OperatingProfitGrowRate")
        self.factor[['ROE','ROA']].plot(figsize=(15, 4),title=u'ROE,ROA')
        self.factor[['EPS','EGRO']].plot(figsize=(15, 4),title=u"EPS,EGRO")
    
    #以三年定期存款利率*2 作为为市场必要回报率估值, 计算市场合理PE(安全PE)
    def market_PE(self):
        marketPE = round(1 / (comm.getRate(self.tdate) * 2)  * 100, 1)
        return marketPE
   
    # 个股历史估值中枢: 个股安全PE
    # 股票的市盈率应小于其过去五年最高市盈率的40% 或 5年市盈率中值 最小值。
    def stock_PE(self):
        PE = self.factor.PE.dropna()
        MAX_PE = np.max(PE) * 0.4
        MID_PE = np.median(PE)
        stock_pe = round(min(MAX_PE,MID_PE),2)
        return stock_pe
    
    #永续债券法求得净资产 PB,假设当前净资产为1)
    def fut_stock_roe_pb(self):
        roe = self.factor.ROE.dropna()
        curr_roe = roe[-1]
        mean_roe = np.mean(roe)
        min_roe = min(mean_roe,curr_roe)
        min_roe =0 if min_roe < 0 else min_roe
        fut_stock_pb = round(min_roe*100/comm.getRate(self.tdate),2)
        return fut_stock_pb
    
    #未来价值
    def intrinsic_value(self):
        fut_pb = self.fut_stock_roe_pb()
        cur_pb = self.factor.PB[-1]
        cur_price = self.his_price.closePrice[-1]
        intrinsic_value = round(cur_price/cur_pb * fut_pb,2)
        return intrinsic_value
    #历史价值: 
    def hist_value(self):
        his_pe = self.stock_PE() 
        cur_pe = self.factor.PE[-1]
        cur_price = self.his_price.closePrice[-1]
        hist_value = round(cur_price/cur_pe * his_pe,2)
        return hist_value
    
    def position(self):
        pos = 0.0
        intrinsic_value = self.intrinsic_value()
        cur_price = self.his_price.closePrice[-1]
        hist_value = self.hist_value()
        bottom_price = min(intrinsic_value,hist_value)
        top_price = max(intrinsic_value,hist_value)
        if bottom_price > 0 and top_price >0 :
            if cur_price < bottom_price :
                pos = 1
            elif cur_price < top_price :
                pos  = round ((top_price - cur_price) /  (top_price - bottom_price),1)
            else :
                pos = 0
        return pos
        						
    def getValue(self):
        value = {}
        value["ticker"] =    self.pcode
        value["secID"] =    self.his_price.secID[-1]
        value["secShortName"] =    self.his_price.secShortName[-1]
        value["tradeDate"] =    self.tdate
        value["市场必要市盈率"] =    self.market_PE()
        value["历史市盈率"] =    self.stock_PE() 
        value["当前市盈率"] =    round(self.factor.PE[-1],2)
        value["预测PB"] =    self.fut_stock_roe_pb()
        value["当前PB"] =    round(self.factor.PB[-1],2)
        value["未来价值"] =    self.intrinsic_value()
        value["当前价格"] =   round(self.his_price.closePrice[-1],2)
        value["历史价值"] =    self.hist_value()
        value["价值位置"] = self.position()
        df = pd.DataFrame(value,index=[self.pcode])
        return df

####### test -----------------------
# stk = StockValue('601985',dt.datetime.strptime('20160405',"%Y%m%d"))
# stk.getValue()
```

```python
#股票趋势计算------------------
class StockTrend:
    pcode = ''
    tdate = dt.date.today()
    his_price = pd.DataFrame()
    factor = pd.DataFrame()
  
    def __init__(self,code,trade_date =dt.date.today()): 
        self.pcode = code
        self.tdate = comm.getPrevBusinessDay(trade_date).strftime("%Y%m%d") 
        self.loaddata()
        
    def loaddata(self):
        #当天个股价格(不复权)
        self.his_price = DataAPI.MktEqudGet(ticker=self.pcode,tradeDate=self.tdate,field=u"",pandas="1")
        self.his_price.index = self.his_price.tradeDate
        
        #当天个股因子数据
        self.factor = DataAPI.MktStockFactorsOneDayGet(ticker=self.pcode,tradeDate = self.tdate,field=['secID','tradeDate','MACD','MA5','MA10'],pandas="1")
        self.factor.index = self.factor.tradeDate
        
    def getTrendByMa(self):
        ma5 = self.factor.MA5[0]
        ma10 = self.factor.MA10[0]
        trend = 100* (ma5 - ma10)/ma5 
        return trend
    def getTrendByMacd(self):
        macd = self.factor.MACD[0]
        return macd
    def getTrend(self):
        trend = self.getTrendDict()['Trend']
        return trend
    
    def getTrendDirect(self):
        trend = "Up" if self.getTrendByMa() > 0.2 and self.getTrendByMacd() > 0 else "Down"
        return trend
    
    def getTrend(self):
        trend = {}
        trend["ticker"] =    self.pcode
        trend["secID"] =    self.his_price.secID[-1]
        trend["secShortName"] =    self.his_price.secShortName[-1]
        trend["tradeDate"] =    self.tdate
        trend["趋势方向"] =   self.getTrendDirect()
        trend["MACD趋势"] =   self.getTrendByMacd()
        trend["MA趋势"] =   self.getTrendByMa()
        df = pd.DataFrame(trend,index=[self.pcode])
        return df
    
# trend = StockTrend('601985',dt.datetime.strptime('20160405',"%Y%m%d"))
# trend.getTrend()
      
```

# 动量指标
-----------
在所读研报中，经过其回测分析，向我们推荐了三大因子：
> CR指标：以上一个计算周期（如N日）的中间价比较当前周期（如日）的最高价、最低价，计算出一段时期内股价的“强弱”，从而在分析一些股价的异常波动行情时，有其独到的功能。另外，CR指标不但能够测量人气的热度、价格动量的潜能，而且能够显示出股价的压力带和支撑带，为分析预测股价未来的变化趋势，判断买卖股票的时机提供重要的参考。

    TYP = (highest + lowest + close)/3;
    CR = sum(max(highest – prev_typical, 0), N) / sum(max(prev_typical – lowest, 0), N) * 100

> PVT指标：价量趋势(PVT)指标是指很巧妙地把能量变化与价格趋势有机地联系到了一起，从而构成了量价趋势指标。参考链接：http://en.wikipedia.org/wiki/Volume-price_trend

    PVT = PVT(-N) + volume * (close - prev_close(N))/prev_close(-N)

> Aroon指标：Aroon是由Tushar Chande发明的，它通过计算自价格达到近期最高值和最低值以来所经过的期间数，帮助投资者预测证券价格从趋势到区域区域或反转的变化。参考链接：http://wiki.mbalib.com/wiki/Aroon
Aroon指标分为Aroon、AroonUp和AroonDown3个具体指标。若考察一段长度为N（一般取26个交易日）的区间，以x表示该区间最高价出现日距离当前交易日的天数，以y表示该区间最低价出现日距离当前交易日的天数。

    AroonUp = 100*(N-x)/N
    AroonDown = 100*(N-y)/N
    Aroon = AroonUp - AroonDown

```python
def CR(high, close, low):
    if len(high) != 21 or len(close) != 21 or len(low) != 21:
        print "history windows length error"
        return
    index = 1
    denominator = 0
    numerator = 0
    while index < 21:
        mid = (high[index] + low[index] + close[index])/3.0
        last_mid = (high[index-1] + low[index-1]+ close[index-1])/3.0
        numerator += max(high[index]- last_mid, 0)
        denominator += max(last_mid - low[index], 0)
        index += 1
    if denominator == 0:
        print "denominator is zero"
        return
    return numerator/float(denominator)

def PVT(account, N):
    temp_dict = account.PVT_record[-N]
    new_dict = {}
    all_close_prices = account.get_attribute_history('closePrice', N+1)
    last_volumes = account.get_attribute_history('turnoverVol', 1)
    for stk in account.universe:
        if stk in temp_dict.keys() and stk in all_close_prices.keys():
            new_dict[stk] = temp_dict[stk] + (all_close_prices[stk][-1] - all_close_prices[stk][-N])*last_volumes[stk][-1]/float(all_close_prices[stk][-N-1])
        elif stk in all_close_prices.keys():
            new_dict[stk] = (all_close_prices[stk][-1] - all_close_prices[stk][-N])*last_volumes[stk][-1]/float(all_close_prices[stk][-N-1])
        else:
            new_dict[stk] = 0
    account.PVT_record.append(new_dict)
    return

def Aroon(account):
    Aroon_up = {}
    Aroon_down = {}
    Aroon_Res = {}
    all_close_prices = account.get_attribute_history('closePrice', N)
    for stk in all_close_prices.keys():
        cp_list = all_close_prices[stk].tolist()
        Aroon_up[stk] = ( N-cp_list.index(max(cp_list)) ) / float(N)
        Aroon_down[stk] = (N-cp_list.index(min(cp_list))) / float(N)
        Aroon_Res[stk] = Aroon_up[stk] - Aroon_down[stk]
    return Aroon_up, Aroon_down, Aroon_Res
```
