# 筹码分布中的Alpha
-----


## 交易行为的非对称性
-----
行为金融学认为，股票的价格并不由股票的内在价值所决定，还很大程度上受投资者主体的行为和心理的影响。换句话说，如果投资者的决策过程存在规律或者特定的话，我们便可以利用这种“人性的弱点”来寻找股票市场的定价错误。
<br>
诺贝尔经济学奖得主卡尼曼曾提出前景理论来描述人们的风险决策行为，其理论有三点结论：处于收益状态时，人们往往小心翼翼、厌恶风险、喜欢见好就收；处于亏损状态时，人们往往会极不甘心，宁愿承受更大的风险来赌一把；白捡的100元所带来的快乐，难以抵消丢失100元所带来的痛苦。
<br>
当前景理论应用在股票市场时，便成了股票市场的处置效应，即：投资者急于卖出盈利的股票，而不愿意卖出亏损的股票；损失股票的持有时间比收益股票的持有时间长；当股票的价格接近投资者的购买价格的时候，投资者效用会发生剧烈改变。
<br>
本文希望通过量化的工具来挖掘处置效应导致的市场定价错误，从而得到不一般基本面分析和资金面分析所不能涉及的Alpha信息，并且我们有理由相信这种基于“人性的弱点”的Alpha是显著而且具有长期有效性的。

## 筹码分布的计算
----
通过上面的分析我们可以知道，若我们知道各个投资者的购买成本，那么我们就可以分析出来股票市场是否真的存在处置效应，并可以以此指导我们进行投资。但是除了交易所，没有人可以知道其他投资者的购买成本的。因此，此处我们根据市场上已有的价格与成交数据近似的代表股票市场投资者的购买成本以及购买数量，以此计算股票的筹码分布。
### 调整换手率
我们利用成均价与换手率来近似计算股票的筹码分布情况。具体的计算细节为：
1. 先确定目标交易日；以表1为例，周五为我们的目标交易日；
2. 换手率的路径依赖调整；以表1为例，周四交易的20%的股份在周五还没有被重新交易的股份为20%×（1-20%）=16%；同理周三交易的20%的股份在周五还没有被重新交易的股份为20%×（1-20%）×（1-20%）=13%；
3. 换手率归一化，根据调整换手率的和对每日的调整换手率进行归一化；
<br>

![](http://odqb0lggi.bkt.clouddn.com/56e7cf9c228e5b858bc25952/6aee5004-99ad-11e6-b055-f8bc124ed898)

可以看出这种筹码分布的计算方法，有一个很强的假设在于每一日的换手平均化到之前的每一天。即按表1来说，周一的成交量，在周二有20%被卖出，周三在周二卖出的基础上再有20%被卖出，周四周五同理。对于周五的20%换手来说，是周四成交的数量的20%加上周三成交的数量但在周四未卖出的20%，一直往前加到周一。
<br>
显然，这离真实的筹码分布还有一定的距离，成交量不可能是简单的平摊到之前的每一天。但没有关系，我们以此为近似筹码分布，继续我们的研究。

### 筹码分布的四个代理变量
给定任意一只股票，假设该股票在t日的成交均价为$P_t$，在t-1日的成交均价为$P_{t-1}$，同理在t-n日的成交均价为$P_{t-n}$。如果我们以t日为目标日计算筹码分布的话，则任意一日$n(1\leq n \leq60)$被交易的股份的相对资产收益（注意：相对资本收益并不是指投资回报）为：

$$RC_{t-n}=\frac{P_{t}-P_{t-n}}{P_{t}}$$
假设股票在t日的换手率为$TR_t$，在t-n日的换手率为$TR_{t-n}$。因此，根据我们上述的换手率调整方法，如果目标日为t日的话，则股票在t-n日的调整换手率为：
$$ATR_{t-n}=TR_{t-n}\cdot(1-TR_{t-n+1})\cdot(1-TR_{t-n+2})\cdots(1-TR_{t})$$
根据相对资本收益以及调整换手率，我们可以构建出筹码分布的均值，方差，偏度和峰度四个代理变量：
$$ARC_{t}=\frac{\sum_{n=1}^{60}ATR_{t-n}\cdot RC_{t-n}}{\sum_{n=1}^{60}ATR_{t-n}}$$
$$VRC_{t}=\frac{n\cdot \sum_{n=1}^{60}ATR_{t-n}\cdot(RC_{t-n}-ARC_{t})^2}{(n-1)\sum_{n=1}^{60}ATR_{t-n}}$$
$$SRC_{t}=\frac{n\cdot \sum_{n=1}^{60}ATR_{t-n}\cdot(RC_{t-n}-ARC_{t})^3}{(n-1)\cdot VRC_{t}^{3/2}\sum_{n=1}^{60}ATR_{t-n}}$$
$$KRC_{t}=\frac{n\cdot \sum_{n=1}^{60}ATR_{t-n}\cdot(RC_{t-n}-ARC_{t})^4}{(n-1)\cdot VRC_{t}^2\sum_{n=1}^{60}ATR_{t-n}}$$
**这四个代理变量的直观含义如下：**
1. ARC>0表示筹码处于平均盈利状态；ARC<0表示筹码处于平均亏损状态；
2. VRC特别大表示筹码特别的分散；VRC特别小表示筹码比较集中；
3. SRC>0表示筹码分布不对称且右偏，即有一部分投资者的收益特别高；SRC<0表示筹码分布不对称且左偏，即有一部分投资者亏损特别严重；
4. KRC特别大表示筹码分布里面盈亏分化很强，要么盈利很多，要么亏损很多，处于小赢小亏状态的筹码非常少；KRC特别小则表现盈亏分化很小，筹码处于小赢小亏的状态；

## 筹码分布的统计特征
-----
我们使用优矿的行情数据，每个月的最后一个交易日计算全部A股的ARC、VRC、SRC、KRC数据。同时我们剔除ST、*ST等有退市风险的股票；剔除在计算日停牌的股票；剔除据计算日上市时间少于240个交易日的股票，作为股票池。

三个工具函数用于筛选股票池

```python
def st_remove(source_universe, st_date=None):
    st_date = st_date if st_date is not None else datetime.datetime.now().strftime('%Y%m%d')
    df_ST = DataAPI.SecSTGet(secID=source_universe, beginDate=st_date, endDate=st_date, field=['secID'])
    return [s for s in source_universe if s not in list(df_ST['secID'])]
```

```python
def secID2ticker(secID):
    """
    给定股票内部编码转化为股票代码
    Args:
        secIDs (list): 需要转化的股票编码列表
    Returns:
        list: 转化为股票代码的列表

    Examples:
        >> secIDs = set_universe('HS300')
        >> tickers = secID2ticker(secIDs)
    """
    universe = DataAPI.EquGet(equTypeCD=u"A",listStatusCD="L,S,DE,UN",field=u"ticker,secID",pandas="1") # 获取所有的A股（包括已退市）
    universe = dict(universe.set_index('secID')['ticker'])
    if isinstance(secID, list):
        res = []
        for i in secID:
            if i in universe:
                res.append(universe[i])
            else:
                print i, ' 在universe中不存在，没有找到对应的secID！'
        return res
    else:
        raise ValueError('secID should be list！')
```

```python
import datetime
from CAL.PyCAL import *
cal = Calendar('China.SSE')

def new_remove(ticker,tradeDate= None,day = 1):
    """
    给定股票列表,去除某个日期前多少个交易日,之后上市的新股
    Args:
        ticker (list of str): 需要进行筛选的股票列表（无后缀）
        tradeDate (datetime): 进行筛选的日期,默认为调用当天
        day (int): 向前漂移的交易日的个数
    Returns:
        list: 去掉新股股票之后的股票列表（无后缀）

    Examples:
        >> universe = set_universe('A')
        >> universe = secID2ticker(universe)
        >> universe_without_new = new_remove(universe)
    """
    tradeDate = tradeDate if tradeDate is not None else datetime.datetime.now()
    period = '-' + str(day) + 'B'
    pastDate = cal.advanceDate(tradeDate,period)
    pastDate = pastDate.strftime("%Y-%m-%d")
    ipo_date = DataAPI.SecIDGet(partyID=u"",assetClass=u"e",ticker=ticker,cnSpell=u"",field=u"ticker,listDate",pandas="1")
    remove_list = ipo_date[ipo_date['listDate'] > pastDate]['ticker'].tolist()
    return [stk for stk in ticker if stk not in remove_list]
```

```python
def get_date_list(headortail = 1):
    ## 可以得到20015年1月至2015年9月交易日里每个月的月初与月末日期list
    calendar_date = DataAPI.TradeCalGet(exchangeCD=u"XSHG",beginDate=u"20050101",endDate=u"20160930",
                                        field=u"",pandas="1")[['isOpen','calendarDate']]
    calendar_date = calendar_date[calendar_date['isOpen']==1]
    calendar_date['calendarDate'] = pd.to_datetime(calendar_date['calendarDate'])
    calendar_date['year'] = calendar_date['calendarDate'].map(lambda x:x.year) 
    calendar_date['month'] = calendar_date['calendarDate'].map(lambda x:x.month)
    if headortail == 1:
        trade_date = calendar_date.groupby(['year','month']).tail(1)['calendarDate'].tolist()
    else:
        trade_date = calendar_date.groupby(['year','month']).head(1)['calendarDate'].tolist()
    return trade_date
```

```python
def get_distribution_of_chips(secID,tradeDate,window):
    
    ## 计算筹码分布的代理变量
    trade_date = tradeDate if isinstance(tradeDate,datetime.datetime) else parse(tradeDate)
    period = '-' + str(window) + 'B'
    begin_date = cal.advanceDate(trade_date,period)
    secID = st_remove(secID,st_date = trade_date.strftime('%Y%m%d'))
    ticker = secID2ticker(secID)
    ticker = new_remove(ticker,tradeDate= trade_date,day = 240)
    
    df_quotation = DataAPI.MktEqudAdjGet(tradeDate=u"",secID=u"",ticker=ticker,isOpen="1",beginDate=begin_date.strftime('%Y%m%d'),
                        endDate=trade_date.strftime('%Y%m%d'),field=u"tradeDate,secID,turnoverValue,turnoverVol,turnoverRate,accumAdjFactor",pandas="1")
    df_quotation['avgPrice'] = df_quotation['accumAdjFactor']*df_quotation['turnoverValue'] / df_quotation['turnoverVol']
    df_quotation = df_quotation[['tradeDate','secID','avgPrice','turnoverRate']]
    df_quotation = df_quotation.set_index(['tradeDate','secID'])
    df_avgPrice = df_quotation.unstack()['avgPrice']
    df_turnoverRate = df_quotation.unstack()['turnoverRate']
    # print df_avgPrice
    # print df_turnoverRate
    df_rc = (df_avgPrice.iloc[-1] - df_avgPrice.iloc[0:-1]) /  df_avgPrice.iloc[-1]
    df_atr = df_turnoverRate.copy()
    for i in range(len(df_atr)):
        df_atr.iloc[i] = df_turnoverRate.iloc[i] * (1 - df_turnoverRate.iloc[i+1:]).prod()
    df_atr = df_atr.dropna(axis=0,how='all')
    arc = (df_atr * df_rc).sum() / df_atr.sum()
    vrc = window*(df_atr*(df_rc - arc)**2).sum()/((window-1)*df_atr.sum())
    src = window*(df_atr*(df_rc - arc)**3).sum()/((window-1)*vrc**(3.0/2)*df_atr.sum())
    krc = window*(df_atr*(df_rc - arc)**4).sum()/((window-1)*vrc**2*df_atr.sum())
    return pd.concat([arc,vrc,src,krc],axis=1).rename(columns={0:'arc',1:'vrc',2:'src',3:'krc'})
```

```python
## 用一个dict收集所有数据，运行时间较长
date_chips_dict ={}
for date in tail_date:
    date_chips_dict[date] = get_distribution_of_chips(set_universe('A'),date,60)
```

```python
def get_factor_df(date_chips_dict,factor_name):
    
    ##传入date_chips_dict，与代理变量名，返回包含三个分位数的dataframe
    factor_dict = {}
    for date in date_chips_dict:
        tmp = date_chips_dict[date][factor_name]
        tmp = tmp.dropna()
        factor_dict[date] = [tmp.quantile(0.25),tmp.quantile(0.5),tmp.quantile(0.75)]
    factor_df = pd.DataFrame(factor_dict).T
    return factor_df
```

```python
## 计算代理变量统计特征
arc_df = get_factor_df(date_chips_dict,'arc').rename(columns={0:'arc_p25',1:'arc_p50',2:'arc_p75'})
vrc_df = get_factor_df(date_chips_dict,'vrc').rename(columns={0:'vrc_p25',1:'vrc_p50',2:'vrc_p75'})
src_df = get_factor_df(date_chips_dict,'src').rename(columns={0:'src_p25',1:'src_p50',2:'src_p75'})
krc_df = get_factor_df(date_chips_dict,'krc').rename(columns={0:'krc_p25',1:'krc_p50',2:'krc_p75'})
```

## 单因子组合分析
我们将对ARC、VRC、SRC和KRC四个因子进行单因子组合分析，来观察者4个因子是否对股票的未来收益有显著的影响能力。此处我们调用Alpalens进行简单分析。首先我们要得到各个股票的开盘价，用于计算每一期的收益，因为每一期的因子都是在月末，用到了当日的成交均价计算的，故对应下一期的收益应该是，下一个月的月初到下下个月的月初开盘价的变动比率。

```python
from lib.alphalen_utils import *
from lib.alphalen_perf import *
from lib.alphalen_plot import *
```

```python
def clean_universe(secID,date):
    ## 清晰A股，剔除ST、新股等
    trade_date = date if isinstance(date,datetime.datetime) else parse(date)
    yesterday = cal.advanceDate(trade_date,'-1B')
    secID = st_remove(secID,st_date = yesterday.strftime('%Y%m%d'))
    ticker = secID2ticker(secID)
    ticker = new_remove(ticker,tradeDate= yesterday,day = 240)
    return ticker
```

```python
def get_open_price(secID,tradeDate):
    ## 获取对应时间点的开盘价
    quotation_list = []
    for date in tradeDate:
        ticker = clean_universe(secID,date)
        one_day_quotation = DataAPI.MktEqudAdjGet(tradeDate=date.strftime('%Y%m%d'),secID=u"",ticker=ticker,isOpen="",
                                                  beginDate=u"",endDate=u"",field=u"secID,tradeDate,openPrice",pandas="1")
        quotation_list.append(one_day_quotation)
    quotation_df = pd.concat(quotation_list,axis=0)
    quotation_df['tradeDate'] = quotation_df['tradeDate'].map(lambda x:parse(x))
    quotation_df = quotation_df.set_index(['tradeDate','secID'])
    return quotation_df
```

```python
def get_cumulative_returns_by_quantile(date_chips_panel,head_tail_dict,factor_name,quotation_df,cut=10):
    
    ##画出分组累计收益图
    factor_ = date_chips_panel[factor_name].iloc[0:-1]
    factor_ = factor_.rename_axis(lambda x:head_tail_dict[x])
    factor, forward_returns = get_clean_factor_and_forward_returns(factor_.stack(), 
                                                               quotation_df.unstack()['openPrice'].iloc[1:].replace(0,np.NaN),periods=[1])
    quantized_factor = quantize_factor(factor,quantiles=cut)
    mean_return_by_monthly, std_err_by_q = mean_return_by_quantile(quantized_factor,forward_returns,by_time='D')
    plot_cumulative_returns_by_quantile(mean_return_by_monthly)
```

## Fama MacBeth回归 
为了研究在控制其他因素不变的情况下，因子的预测能力是否会发生显著改变。我们使用经典的面板数据回归方法来看因子的影响能力是否显著。

```python
def get_fama_reg_df(quotation_df,date_chips_panel,head_tail_dict):
    ##整合数据，形成用于回归的dataframe
    def get_rule_factor_df(factor_name,head_tail_dict):
        factor = date_chips_panel[factor_name].iloc[0:-2]
        factor = factor.rename_axis(lambda x:head_tail_dict[x])
        factor_df = pd.DataFrame(factor.stack())
        factor_df.index.names = ['tradeDate','secID']
        factor_df = factor_df.rename(columns={0:factor_name})
        return factor_df
    
    fama_reg_df = quotation_df.unstack().replace(0,np.NaN).pct_change().shift(-1).iloc[1:-1].stack()
    fama_reg_df = fama_reg_df.rename(columns={'openPrice':'equityRet'})
    factor_arc_df = get_rule_factor_df('arc',head_tail_dict)
    factor_vrc_df = get_rule_factor_df('vrc',head_tail_dict)
    factor_src_df = get_rule_factor_df('src',head_tail_dict)
    factor_krc_df = get_rule_factor_df('krc',head_tail_dict)
    return pd.concat([fama_reg_df,factor_arc_df,factor_vrc_df,factor_src_df,factor_krc_df],axis=1)
```

```python
fama_reg_df = get_fama_reg_df(quotation_df,date_chips_panel,head_tail_dict)
```

回归的例子如上，我们将所有回归汇总成表，如下所示。

```python
reg_result = OrderedDict((('arc_reg',[fm_list[0].mean_beta[1],fm_list[0].mean_beta[0],'-','-','-']),
('t_stat_arc',[fm_list[0].t_stat[1],fm_list[0].t_stat[0],'-','-','-']),
('vrc_reg',[fm_list[1].mean_beta[1],'-',fm_list[1].mean_beta[0],'-','-']),
('t_stat_vrc', [fm_list[1].t_stat[1],'-',fm_list[1].t_stat[0],'-','-']),
('src_reg',[fm_list[2].mean_beta[1],'-','-',fm_list[2].mean_beta[0],'-']),
('t_stat_src',[fm_list[2].t_stat[1],'-','-',fm_list[2].t_stat[0],'-']),
('krc_reg',[fm_list[3].mean_beta[1],'-','-','-',fm_list[3].mean_beta[0]]),
('t_stat_krc',[fm_list[3].t_stat[1],'-','-','-',fm_list[3].t_stat[0]]),
('all_reg',[fm_list[4].mean_beta[4],fm_list[4].mean_beta[0],fm_list[4].mean_beta[1],fm_list[4].mean_beta[2],fm_list[4].mean_beta[3]]),
('t_stat_all',[fm_list[4].t_stat[4],fm_list[4].t_stat[0],fm_list[4].t_stat[1],fm_list[4].t_stat[2],fm_list[4].t_stat[3]])))
pd.DataFrame(OrderedDict(reg_result)).T.rename(columns={0:'Intercept',1:'ARC',2:'VRC',3:'SRC',4:'KRC'})
```

从上表的回归结果来看，单因子回归中，KRC是最显著的，SRC是最不显著的；从方向来看，ARC和KRC是反向因子，即因子值越小，股票未来收益越高，VRC和SRC是正向因子，即因子值越大，股票未来收益越高；
<br>
从最后的多因子回归的的结果来看，ARC与VRC的有效性降低，因为这两个因子的相关性很高，与我们相关性分析的结果相符合；VRC因子的系数由正转负，因为VRC在ARC大于0和小于0时存在相关性的结构改变；SRC没有太大变化；KRC因子的有效性提升，这意味着KRC因子隐含着筹码分布里面很丰富的信息。

从上述的回归结果来看，我们反向在Updata与Downdata回归方程里面ARC和VRC的系数与显著程度都截然不同。在Updata里面，股票的下期收益与VRC和KRC呈现显著的负相关性。换句话说，在Updata里面，处于筹码平均盈利的，筹码分布越集中，盈亏分化较小的股票以后会获得更高的收益。
<br>
在Downdata中,股票收益与VRC呈现出显著的正相关性,即在Downdata里面，处于筹码平均亏损的，筹码分布发散的的股票以后会获得更高的收益。
<br>
与Alldata的回归结果比较的话，我们可以发现，通过将全样本划分为Updata和Downdata，我们将VRC的显著性得到极大的提升，结构变化相当明显。

### 盈亏划分的因子重构
基于上述的分析结果，我们知道在arc_p50大于0的时候，VRC与下期收益呈现显著的负相关性；在arc_p50小于0的时候，VRC与下期收益呈现出显著的正相关性。故我们在arc_p50大于0时，取VRC的负值，在arc_p50小于0时，取VRC的原值，得到重构的VRC因子，如下所示。
