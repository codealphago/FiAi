## 策略流程：
---

1. __选股__

    网格策略的逻辑是在股票便宜时买进，估值过高时卖出，因此选出波动率较高的股票很有必要，因此本篇的选股流程如下：  
    
    先设置股票池行业为中证行业：ZZ500 或 CYB , 或其他行业指数, 信息产业, 软件应用, 通信, 大数据, 新能源新技术；  
    
    估值不能过高：0< PE < 60 , ROE > 0.07；  
    
    市值约束：取满足上述条件的市值较小的30只股票；  
    
    高波动：分行业在市值最小的30只中选出过去一年波动率最大的5-10只股票；  
    
    上述流程后，我们有了5-10只股票构成的股票池，每隔60个交易日更新一次股票池。

2. __网格__ 

    三种大小的网格都会相应尝试一下看看效果。
	优化: 改成根据本次选股的波动进行网格大小的动态调整范围:[0.03-0.08]
    买入卖出各 4档 相同大小网格
    
    [-3%，3%卖]、[-5%买，5%卖]、[-8%买，8%卖]

3. __资金安排__

    在仓位控制时，满仓的概念是（总资金/股票池总数*1.5），这是为了提高资金利用率，因为3个月的周期内可能不是每只股票都能达到满仓。

4. __止损__
    
    (基准前两日亏损3%以上则空仓) : 当前条件不执行.
    个股累计亏损30%则平仓该股。

---
1. 对于三种不同网格，在不同市场时期回测，得到结果：
2. __震荡期用小网格__：`2012/01/01 - 2014/11/01`这段震荡时期中，小网格策略实现48.3%的年化收益，同期基准收益0.8%. 看来市场震荡期网格操作大有用武之地；
3. __慢趋势期用中网格__：`2015/08/31 - 2016/08/01`这段慢熊时期中，中网格策略实现50%年化收益，同期基准年化收益-5.5%. 这是因为市场有一定趋势时，网格大有一定容错率，否则容易过快实现亏损，得不到收益；
4. __长期投资用大网格__:`2010/01/01 - 2016/08/01`这段时间，大网格实现15.5%的年化收益，同期基准年化收益为-0.1%. 同样的逻辑，时间越长，市场越可能走成趋势，需要提高容错率；
5. __以上收益的来源并不来自选股/行业，而是来自操作方法。__在`2010/01/01 - 2016/08/01`这段时间，仍按上述方法选股、止损，但只等额持有股票，不做网格操作，年化收益只有6.6%，而最大回撤高于60%，表现并不理想。

```python
import pandas as pd
import numpy as np
import lib.grid as grid
import lib.comm as comm
import datetime as dt

start = '2015-01-01'                       # 回测起始时间
end = '2016-09-28'                         # 回测结束时间
#today = comm.getPrevBusinessDay(dt.date.today(),1)
#end = today.strftime('%Y-%m-%d') 
benchmark = 'HS300'                        # 策略参考标准
universe = DynamicUniverse('HS300') #DynamicUniverse(IndZZ.XinXiJiZhuL1) +  DynamicUniverse(IndZZ.DianXinYeWuL1)  #DynamicUniverse('HS300') ## 证券池，支持股票和基金
capital_base = 1000000                      # 起始资金100W
freq = 'd'                                 # 策略类型，'d'表示日间策略使用日线回测，'m'表示日内策略使用分钟线回测
refresh_rate = 1                         # 调仓频率，表示执行handle_data的时间间隔，若freq = 'd'时间间隔的单位为交易日，若freq = 'm'时间间隔为分钟

# ---------------回测参数部分结束----------------

# 把回测参数封装到 SimulationParameters 中，供 quick_backtest 使用
sim_params = quartz.SimulationParameters(start, end, benchmark, universe, capital_base)
# 获取回测行情数据
data = quartz.get_backtest_data(sim_params)
# 运行结果
cumulative_returns = {}
annualized_return = {}
max_drawdown = {}
# 调整参数快速回测
for x in [3,4,5,6,7,8] :# range(1, 10):
    # refresh_rate 的值和回测行情数据无关，所以也可以在参数调优中修改
    # 类似的还有 commission 和 slippage
    # refresh_rate = 2
    # commission = ...
    # slippage = ...

    # 策略特有参数, 网格大小, 资本倍数: 
    grid_size = x * 0.01
    cap_ratio = 1.5
    stock_num = 5
    his = 60
    
    def initialize(account):   
        print "start test " , x
        account.stock_pool = []        # 股票池
        account.base_price = {}        # 股票基准价
        account.adjust_stock_pool = 0  # 股票池调整计数器
        account.adjust_base = 0        # 股票基准价调整计数器
        account.update_windows = 60 / refresh_rate  # 股票池调整频率

    def handle_data(account):
        account.adjust_base += 1
        account.market_adj = False     # 是否根据大盘调整
        account.adjust_stock_pool += 1
        if  account.stock_pool == [] or account.adjust_stock_pool >  account.update_windows:
            account.stock_pool,features = stock_filter(account.previous_date, account.universe,stock_num,his)
            #print "调整股票池：", account.current_date,account.stock_pool
            #print features
            account.grid_size = grid_size
            #min(max(np.round(min(features.vol)-0.01,2),0.03),0.08)
            #print "平均波动率: ",account.current_date, account.grid_size 
            account.adjust_stock_pool = 0
            for stk in account.security_position :
                if stk not in account.stock_pool:
                    #print "调仓卖出股票:" , stk
                    order_to(stk, 0)

        for stk in account.stock_pool:        
            if (not stk in account.reference_price) or account.reference_price[stk] == 0:
                continue
            # 获取 股票时间序列数据
            stk_his = grid.get_his_return(account,stk,20)
            # 加入基准价    
            if not stk in account.base_price:
                account.base_price[stk] = account.reference_price[stk] 

            adj_pct = account.reference_price[stk] / account.base_price[stk]
            max_pct = 1.00 / len(account.stock_pool) * cap_ratio # 单只股票最大仓位
            # 根据网格调仓
            pct = grid.adj_positon(adj_pct,max_pct,account.grid_size,account.grid_size)
            if pct >= 0 : 
                order_pct_to(stk,pct)
        return
# ---------------策略逻辑部分结束----------------

    # 把回测逻辑封装到 TradingStrategy 中，供 quick_backtest 使用
    strategy = quartz.TradingStrategy(initialize, handle_data)
    bt, perf = quartz.quick_backtest(sim_params, strategy, data=data, refresh_rate=refresh_rate)
    # 保存运行结果
    cumulative_returns[x] = perf['cumulative_returns']
    annualized_return[x] = perf['annualized_return']
    max_drawdown[x] = perf['max_drawdown']

print "Finished..."
```
