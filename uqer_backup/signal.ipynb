市场(指数)换手率信号, 
如果当前换手率 > 10日移动平均 或者 当前换手率 > 20 日移动平均 并且 当天指数下跌不超过2% 则是买入信号 
否则当前换手率 < 10日移动平均 并且 当前换手率 < 20 日移动平均 或者 当天指数下跌超过2%   则是卖出信号 

```python
# 本市场活跃度（人气）： 每月日均换手率，分母为上证A股流通市值, 反映市场人气, 人气高则指数还可以继续维持原来趋势或者容易创出新高,
# startDate , endDate type is dt.date
def getIndexTurnover(indexID, startDate,endDate):
    sDate =  startDate.strftime('%Y%m%d')
    trade_days = comm.getBusinessDays(startDate,endDate)
    mkt_value  = [] 
    stock_list = []
    if indexID == 'MKT' :
        stock_list = '' 
    else :
        stock_list = DataAPI.IdxConsGet(ticker=indexID,intoDate=sDate,isNew=u"",field=u"consID",pandas="1").consID.values
    for tradeDate in trade_days :
        data_temp = DataAPI.MktEqudGet(secID = stock_list , tradeDate=tradeDate,field=u"turnoverValue,negMarketValue",pandas="1")
        total = data_temp.sum()
        mkt_value.append([tradeDate,total.turnoverValue,total.negMarketValue])
    mkt = pd.DataFrame(mkt_value,columns=['tradeDate','turnoverValue','negMarketValue']) 
    mkt.index = mkt.tradeDate
    del mkt['tradeDate']
    mkt['turnoverRate'] = 100 * mkt.turnoverValue / mkt.negMarketValue 
    mkt = mkt.dropna()
    return mkt 

#  市场成交量人气指数
#  需要计算22日移动平均 所以 n > 22
def market_activity(indexID,tdate,n=25):

    sdate = comm.getPrevBusinessDay(tdate,n)
    activity = getIndexTurnover(indexID,sdate,tdate)
    if indexID == 'MKT':   
        indexID = '000001'
    mkt = DataAPI.MktIdxdGet(ticker=indexID,beginDate=sdate,endDate=tdate,field=['tradeDate','closeIndex','CHGPct'],pandas="1")
    activity['roll10'] = pd.rolling_mean(activity.turnoverRate,10)
    activity['roll20'] = pd.rolling_mean(activity.turnoverRate,22)
    mkt.index = mkt.tradeDate
    data = pd.concat([activity,mkt],axis=1)
    data = data.dropna()
    return data

def display(data):    
    fig = plt.figure() # Create matplotlib figure
    ax = fig.add_subplot(111) # Create matplotlib axes
    ax2 = ax.twinx() # Create another axes that shares the same x-axis as ax.
    # roll_rate 是 22个交易日滚动平均数（量*价/流通市值）
    data.plot(y=['turnoverRate','roll20','roll10'], kind='line',ax= ax , figsize=(18, 4), title=u'成交量人气指数')
    data.plot(y=['closeIndex'],kind='line',ax= ax2, figsize=(18, 5),color='r')
    report.fig_style(ax, [u'换手率',u'22日平均还手率',u'10移动平均还手率'] , 'upper left')
    report.fig_style(ax2, [ u'指数价格'] )

def getActivateSignal(indexID,tdate=dt.date.today(),n=25):
    activity = market_activity(indexID,tdate,n)
    activity['diff10'] = activity.turnoverRate - activity.roll10
    activity['diff20'] = activity.turnoverRate - activity.roll20  
    activity['sig10'] = False
    activity['sig20'] = False
    activity['sig'] = False
    for idx in activity.index: 
        sig = activity.loc[idx].diff10 > 0 or activity.loc[idx].diff20 > 0 
        sig10 = activity.loc[idx].diff10 > 0 
        sig20 = activity.loc[idx].diff10 > 0 
        if activity.loc[idx].CHGPct < -0.02 : 
            sig20 = sig = sig10 = False
        activity.at[idx,'sig'] = sig
        activity.at[idx,'sig10'] = sig10
        activity.at[idx,'sig20'] = sig20
    return activity
    
# print getIndexTurnover(CYB,dt.date(2016,12,1),dt.date(2016,12,22))    
#activity = market_activity(CYB,dt.date(2016,11,24))
# sig1 = activity.diff10[-1] > 0 or activity.diff20[-1] > 0 
# sig2 = activity.diff10[-2] > 0 or activity.diff20[-2] > 0
# print sig1 and sig2
# df = activity.tail(20)
# df
# activity = getActivateSignal(CYB,dt.date.today(),n=100)    
# display(activity)  
# activity.tail(5)
```

在传统的金融意义上，由于流动性的总量是一定的，而且风险偏好有着巨大的差异，一般认为股票市场和债券市场存在着跷跷板效应，股市和债市可以参考着看，有一定的借鉴意义。不过对于中国市场而言，股市和其他市场的关系一般没有那么紧密，所以利用Shibor来进行择时也只是姑且一试。

利用Shibor对股市进行择时的 intuition 即如同上述所说的，当利率处于高位继而下行时，流动性逐渐从货币市场和债券市场释放出来进入股市，会推高股指；反之，则意味着整个资本市场流动性降低，投资者逐渐转向低风险资产，

策略细节方面，首先是Shibor期限的选取，作为择时信号显然要对市场变化敏感，故我选择了7天的Shibor作为信号源，因为其每天的波动过于剧烈，所以对其进行了三个月的平滑。

其次是对于Shibor高低位的判断，这次我尝试了分位点，即以过去120个交易日的Shibor作为经验分布，求出其一定的分位点作为阈值来和调仓日进行比较，我尝试了多种不同的分位点。

再然后是标的的选择，我尝试了沪深300、中证500和创业板指，代码示例中采用的是创业板指，因为表现实在是太抢眼了，为此我反复检查了三四遍确定了没有未来数据。

最后是多空的控制，虽然目标是高位看多低位看空，但还存在着中间情况的可能，对于这种情况，我选择直接看多，因为总是持有指数，即使表现不好，那么也是持仓与标的表现一样不好，在择时的意义上并没有额外的损失。

```python
# shibor 大盘择时
def shibor_stat(term,sdate,edate=dt.date.today(),dist_window = 120, ma_window=60):
    shibor_name = 'shibor_{}'.format(term)
    # dist_window = 120  # 计算分布的时间窗口 , 交易日(半年)
    # ma_window = 60  # 计算平滑的时间窗口 交易日(3个月)
    # 处理分位点和平滑
    shibor_df = comm.get_shibor(term, sdate, edate)
    shibor_df['1/3q'] = pd.rolling_quantile(shibor_df[shibor_name], dist_window, 0.33)
    shibor_df['1/2q'] = pd.rolling_quantile(shibor_df[shibor_name], dist_window, 0.5)
    # shibor_df['2/3q'] = pd.rolling_quantile(shibor_df[shibor_name], dist_window, 0.67)
    shibor_df['ma'] = pd.rolling_mean(shibor_df[shibor_name], ma_window)
    shibor_df = shibor_df.dropna()
    shibor_df.index = shibor_df.tradeDate
    shibor_df['signal_ma'] =  shibor_df.ma  >= shibor_df[shibor_name]
    shibor_df['signal_dist'] = shibor_df.ma >= shibor_df['1/3q'] 
    shibor_df['signal'] = shibor_df['signal_ma'] & shibor_df['signal_dist']
    return shibor_df

def getShiborSignal(term,tdate=dt.date.today(),window=60):
    sdate = tdate - dt.timedelta(365)
    shibor_df = shibor_stat('1W',sdate,edate=tdate,ma_window=window)
    signal = shibor_df
    return signal


# getShiborSignal('1W')
# shibor_df = shibor_stat('1W',dt.date(2015,1,1),edate = dt.date(2016,12,30))
# shibor_df
# shibor_df.tail(10)
# shibor_df.loc[:, ['ma', '1/3q', '1/2q', '2/3q']].plot(figsize=(16,5)) 
```
