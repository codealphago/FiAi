# 公共使用函数库
- getPrevBusinessDay(trade_date,days=0): 得到前n个市场交易日
- getNextBusinessDay(trade_date,days=0): 得到后那个交易日  
- getBusinessTermMove(trade_date,term='1Y'): 得到第N term 之前或之后的交易日 
- getRate(trade_date = dt.date.today(),vType = 'bond'):  M120000047 五年期国债收益率 （bond） # M120000009 3年定期存款利率 （cd）
- his_market_PE(startDate,endDate): 历史市场必要市盈率
- getMarketTotalValue(startDate,endDate): 取得某段时间A股总市值和总流通市值
- getUSDRMB( startDate,endDate,moneyPair = 'USD/CNY'): 取得某段时间 市场人民币汇率 
- getShibor( startDate,endDate): 取得某段时间 市场利率(shibor1D, shibor1W, 交易所回购利率GC001,GC007) 
- getRepo(startDate,endDate): 交易所国债回购利率
- getHKSHdata(startDate, endDate): 一段时间沪港通净流入流出数据
- getMarketStat(startDate,endDate): 市场活跃度（人气）： 每月日均换手率，分母为上证A股流通市值 
- dailyHKvolume(date): 得到某日沪股通的流入流出金额  
- dailySHvolumn(date): 得到某日沪股通的流入流出金额  
- marketPE(startDate, endDate): 取得市场PE中值

```python
def ticker2SecID(ticker):
    return ticker + '.XSHG' if ticker[0:1] == '6' else ticker + '.XSHE'
# print ticker2SecID('012001')

# 得到前n个市场交易日
def getPrevBusinessDay(trade_date,days=0):
    last_day = cal.advanceDate(trade_date,'-'+str(days)+'B',BizDayConvention.Preceding)            #计算出倒数第一个交易日
    return last_day.toDateTime()
# 得到后那个交易日
def getNextBusinessDay(trade_date,days=0):
    last_day = cal.advanceDate(trade_date,str(days)+'B',BizDayConvention.Preceding)            #计算出倒数第一个交易日
    return last_day.toDateTime()

# 得到第N term 之前或之后的交易日
def getBusinessTermMove(trade_date,term='1Y'):
    last_day = cal.advanceDate(trade_date,term,BizDayConvention.Preceding)            #计算出倒数第一个交易日
    return last_day.toDateTime()
# 得到一个范围内的交易日
def getBusinessDays(startDate,endDate):
    data = DataAPI.TradeCalGet(exchangeCD=u"XSHG", beginDate=startDate, endDate=endDate, field=['calendarDate', 'isOpen'])
    data = data[data.isOpen == 1]
    trade_days = data.calendarDate
    return trade_days
# 得到每月最后一个交易日
def getBusinessMonthEndDays(startDate='2010-01-01',endDate=''):
    data = DataAPI.MktIdxdGet(ticker=u"000001",beginDate=startDate, endDate=endDate,field=u"tradeDate",pandas="1")
    data['month'] = data.tradeDate.apply(lambda x : x[0:7]) 
    grpd = data.groupby('month')
    monthEnd = grpd.max()
    return monthEnd.tradeDate.values
#getBusinessMonthEndDays()
# print getBusinessDays("20160301","20160401")
# print getPrevBusinessDay(dt.date.today(),4)
# print getNextBusinessDay(dt.date.today(),4)
# print getBusinessTermMove(dt.date.today(),'-1M')

```

```python
# M120000047 五年期国债收益率 （bond）
# M120000009 3年定期存款利率 （cd）
def getRate(trade_date = dt.date.today(),vType = 'bond'): 
    indic = 'M120000047' if vType == 'bond' else 'M120000009'
    macro_data =  DataAPI.ChinaMacroDataGet(indicID=indic,indicName=u"",beginDate=u"",endDate=trade_date.strftime('%Y%m%d'),field=u"",pandas="1")
    return macro_data.dataValue[0]  
# print getRate(dt.datetime.strptime('20160216','%Y%m%d'),'cd')
```

```python
# 历史市场必要市盈率
def his_market_PE(startDate,endDate):
    macro_data = DataAPI.ChinaMacroDataGet(indicID=u"M120000047",indicName=u"",beginDate=startDate,endDate=endDate,field=u"",pandas="1")
    macro_data.index = macro_data.publishTime
    marketPE =  1 /(macro_data.dataValue *2 ) *100 
    marketPE = marketPE.sort_index()
    return marketPE
#his_market_PE('20160101','20160330')
```

```python
#取得某段时间A股总市值和总流通市值
def getMarketTotalValue(startDate,endDate):
    trade_days = getBusinessDays(startDate,endDate)
    mkt_value = []
    for tradeDate in trade_days :
        #data_temp = DataAPI.MktEqudLatelyGet(secID='',field=u"marketValue,negMarketValue",pandas="1")
        data_temp = DataAPI.MktEqudGet(tradeDate=tradeDate,field=u"marketValue,negMarketValue",pandas="1")
        total = data_temp.sum()
        mkt_value.append([tradeDate,total.marketValue,total.negMarketValue])

    mkt = pd.DataFrame(mkt_value,columns=['tradeDate','marketValue','negMarketValue']) 
    mkt.index = mkt.tradeDate
    del mkt['tradeDate']
    data = DataAPI.FstTotalGet(exchangeCD=u"XSHE,XSHG", beginDate=startDate,endDate= endDate)
    data = data.groupby(by='tradeDate').sum()
    data = pd.concat([data, mkt], axis=1)
    data = data.dropna()
    data['valPct'] = 100 * data['tradeVal']/data['negMarketValue']
    data['tradeVal'] = data['tradeVal'] / (10**12)
    data['tradeValChg'] = data.tradeVal.diff()*(10**4)
    return data 

#getMarketTotalValue('20160325','20160330')
    
```

```python
# 取得某段时间 市场人民币汇率
def getUSDRMB( startDate,moneyPair = 'USD/CNY'):    
    rmb_rate =DataAPI.MktFxRefRateGet(currencyPair=moneyPair,beginDate=startDate,endDate='',field=u"",pandas="1")
    rmb_rate = rmb_rate.sort(['tradeDate'])
    rmb_rate.index = rmb_rate.tradeDate
    del rmb_rate['tradeDate']
    rmb_rate['chg'] =  10000 * rmb_rate.midRate.diff()
    return rmb_rate
# getUSDRMB('20161029')
```

```python
# 取得某段时间 市场利率(shibor1D, shibor1W, 交易所回购利率GC001,GC007)
def getShibor( startDate):
    data = DataAPI.ChinaDataInterestRateSHIBORGet(indicID=u"M120000057,M120000058",indicName=u"SHIBOR",beginDate=startDate,endDate=u"",field=u"indicID,indicName,periodDate,dataValue",pandas="1")
    data = data.set_index(data.periodDate)
    data = data.sort()
    shibor1d = data[data.indicID == 'M120000057'] 
    shibor1w = data[data.indicID == 'M120000058']
    shibor = pd.DataFrame()
    shibor['shibor1D'] = shibor1d.dataValue
    shibor['shibor1W'] = shibor1w.dataValue
    shibor['shibor1dChg']= 100 * shibor.shibor1D.diff()
    shibor['shibor1wChg']= 100 * shibor.shibor1W.diff()
    return shibor

# term = ['1D','1W','2W','1M'] , endDate 延迟一天
def get_shibor(term, start, end=dt.date.today()):
    shibor_df = DataAPI.ChinaDataInterestRateSHIBORGet(indicName="SHIBOR_SH00{}".format(term), beginDate=start, endDate=end, field="periodDate,dataValue").sort_values("periodDate")
    shibor_df.reset_index(inplace=True, drop=True)
    rename_vars = {'periodDate': 'tradeDate',
              'dataValue' :'shibor_{}'.format(term) }
    shibor_df = shibor_df.rename(columns=rename_vars)
    return shibor_df

# get_shibor('1W',dt.date(2016,10,1))
# get_shibor('1M',dt.date(2016,10,1))
# getShibor('20161029')
```

```python
#交易所国债回购利率
def getRepo(startDate):
    repo = DataAPI.MktRepodGet(secID=u"204001.XSHG,204007.XSHG",beginDate=startDate,endDate='',field=u"secShortName,tradeDate,closeRate",pandas="1")
    snapshot = DataAPI.MktTickRTSnapshotGet(securityID=u"204001.XSHG,204007.XSHG",exchangeCD="",assetClass="",field=u"lastPrice,shortNM,dataDate",pandas="1") 
    
    repo = repo.append({'secShortName':snapshot.shortNM.values[0].strip(), 'tradeDate':snapshot.dataDate.values[0], 'closeRate':snapshot.lastPrice.values[0]}, ignore_index=True)
    repo = repo.append({'secShortName':snapshot.shortNM.values[1].strip(), 'tradeDate':snapshot.dataDate.values[1], 'closeRate':snapshot.lastPrice.values[1]}, ignore_index=True)
    repo.index = repo.tradeDate
    repo1d = repo[repo.secShortName == 'GC001']
    repo1w = repo[repo.secShortName == 'GC007']
    
    marketRepo = pd.DataFrame()
    marketRepo['repo1D'] = repo1d.closeRate
    marketRepo['repo1W'] = repo1w.closeRate
    marketRepo['repo1dChg']= 100 * marketRepo.repo1D.diff()
    marketRepo['repo1wChg']= 100 * marketRepo.repo1W.diff()
    marketRepo = marketRepo[marketRepo.repo1dChg <> 0 ]
    marketRepo = marketRepo.dropna() 
    return marketRepo

# df = getRepo('20161029')
# df
```

```python
#一段时间沪港通净流入流出数据
def getHKSHdata(startDate, endDate):
    trade_days = getBusinessDays(startDate,endDate)
    
    res, sh, hk = pd.DataFrame(), pd.DataFrame(), pd.DataFrame()
    sh = DataAPI.equSHHKQuotaGet(exchangeCD=u"SHSC", tradeDate=trade_days, field=u"")
    hk = DataAPI.equSHHKQuotaGet(exchangeCD=u"XSSC", tradeDate=trade_days, field=u"")
    
    sh['netIn'] = (sh['dailyQuota'] - sh['dailyBalance']) / (10 ** 8)
    hk['netIn'] = (hk['dailyQuota'] - hk['dailyBalance']) / (10 ** 8)
    res['tradeDate'], res['in_sh'] = sh['tradeDate'], sh['netIn']
    res = res[res.tradeDate.isin(hk.tradeDate.values)]                
    res['in_hk'] = hk[hk.tradeDate.isin(sh.tradeDate)].netIn   
    
    res.tradeDate = [i[:10] for i in res.tradeDate.values] 

    return res

# getHKSHdata('20160328','20160413') 
```

```python
# 本市场活跃度（人气）： 每月日均换手率，分母为上证A股流通市值, 反映市场人气, 人气高则指数还可以继续维持原来趋势或者容易创出新高,
def getMarketStat(startDate,endDate):
    trade_days = getBusinessDays(startDate,endDate)
    mkt_value  = [] 
    for tradeDate in trade_days :
        data_temp = DataAPI.MktEqudGet(tradeDate=tradeDate,field=u"turnoverValue,negMarketValue",pandas="1")
        total = data_temp.sum()
        mkt_value.append([tradeDate,total.turnoverValue,total.negMarketValue])
    mkt = pd.DataFrame(mkt_value,columns=['tradeDate','turnoverValue','negMarketValue']) 
    mkt.index = mkt.tradeDate
    del mkt['tradeDate']
    mkt['turnoverRate'] = 100 * mkt.turnoverValue / mkt.negMarketValue 
    mkt = mkt.dropna()
    return mkt 

# kt = getMarketStat('20160101','20160331')
# kt.tail(5)
```

```python
#取得市场PE中值
def marketFactor(factors = ["PE","ROE"],startDate='2015-01-01', endDate=''):
    pe = []
    trade_days = getBusinessMonthEndDays(startDate,endDate)  
    for date in trade_days:  
        data_temp = DataAPI.MktStockFactorsOneDayGet(tradeDate = date ,secID='',field=["tradeDate"] +  factors,pandas="1")
        data_temp = data_temp.dropna()
        pe.append([date,data_temp[factors[0]].median(), data_temp[factors[1]].median()])
    df = pd.DataFrame(pe,columns=["tradeDate"] +  factors) 
    df.index = df.tradeDate
    return df

#marketFactor()
```

```python
# 状态突破， 根据HMM 计算大哥中国A股指数状态转换周期为1.6周， 所以当价格突破2个月均线 + 2个月波动率*0.8 以后买入 ，否则卖出 ， 每周重新计算 
def cal_updown_Status(hist_p, k=0.8, p=0.8):
    avg = np.mean(hist_p)
    std = np.std(hist_p)
    x = avg + std*k
    z = avg - std*p
    c = 0.0
    status = 0
    if len(hist_p) > 0 :
        if hist_p[-1] > x :
            status = 1 
        if hist_p[-1] < z : 
            status = -1
        c = hist_p[-1]
    return round(x,2),round(z,2),round(c,2),status
```

```python
# ticker 支持股票， beginDate 要比实际往前推250个交易日
def moneyInfo_data(ticker, beginDate, endDate):
    #使用 基础指标（每日收盘价，每日资金流入，每日资金流出）构造出特征向量（日收益率，资金流入占当日所有流动资金比例，日总流动资金环比，（日总流动资金-过去一年平均日流动资金）/过去一年流动资金波动率
    data = DataAPI.MktEqudAdjGet(ticker=ticker,beginDate=beginDate,endDate=endDate,field=['tradeDate','closePrice','lowestPrice','highestPrice','turnoverVol'])
    data1 = DataAPI.MktEquFlowGet(tradeDate=u"",secID=u"",ticker=ticker,beginDate=beginDate,endDate=endDate,field=u"secID,ticker,secShortName,tradeDate,moneyInflow,moneyOutflow,netMoneyInflow")
    data1 = data1.sort(['tradeDate'])
    
    #构造出特征向量（日收益率，资金流入占当日所有流动资金比例，日总流动资金环比，（日总流动资金-过去一年平均日流动资金）/过去一年流动资金波动率
    data['Return1'] = data.closePrice.pct_change()
    
    data1['totalFlow'] = data1.moneyInflow + data1.moneyOutflow
    data1['inflowPct']=data1.moneyInflow / data1.totalFlow 
    data1['totalPct'] = data1.totalFlow.pct_change()
    data1['yearMean'] = pd.rolling_mean(data1.totalFlow,250)
    data1['yearVol'] = pd.rolling_std(data1.totalFlow,250)
    data1['flowVol'] = (data1.totalFlow - data1.yearMean )/data1.yearVol
    
    res = pd.merge(data,data1,on='tradeDate')
    res = res.dropna()
    res = res.replace('inf',0.0)
    res = res.set_index('tradeDate')
    return res
#### testing ###############
#info = moneyInfo_data('000001','20130101','20160601')
#info.flowVol.hist(bins=50)
#info.totalPct.hist(bins=50)
#info.inflowPct.hist(bins=50)
#info.Return1.hist(bins=50)
```

```python
# 日期格式处理函数
def itoDateStr(year = 2016, month = 6, day = 26):
    return str(year) + ('0'+str(month) if (month<10) else str(month))+('0'+str(day) if (day<10) else str(day))


def toDateTime(argStr = u"2012-01-01"):
    return datetime.date(string.atoi(argStr[0:4]), string.atoi(argStr[5:7]), string.atoi(argStr[8:10]))


# 获取连续期货价格函数
def getFutuCon(bDate = u"20120101", eDate = u"", conObj = u"AU", period = 1):
    histPrice = {}
    if eDate == "":
        eDate = time.strftime("%Y%m%d",  time.localtime(time.time()))
    calendar = DataAPI.TradeCalGet(exchangeCD = u"XSHE", beginDate = bDate, endDate = eDate, field = u"calendarDate,isOpen", pandas = "1")
    calArray = calendar[calendar['isOpen'] == 1]['calendarDate'].values
    
    for date in calArray:
        year = string.atoi(date[2:4]);month = string.atoi(date[5:7])
        year = (year if (month+period<= 12) else year+1)
        month = (month+period if (month+period<= 12) else month+period-12)
        year = str(year)
        month = ('0'+str(month) if(month<10) else str(month))
        tker = conObj+year+month
        priceVal = DataAPI.MktFutdGet(tradeDate = toDateTime(date), ticker = tker, field = u"closePrice")
        histPrice[toDateTime(date)] = (np.NaN if len(priceVal) == 0 else priceVal)
    return pd.DataFrame(histPrice, index = [conObj+'_Continuous_'+str(period)]).T

# fut = getFutuCon(bDate = '20160630', eDate = '20160630', conObj = 'IF', period = 1).values[0][0]
# curr = DataAPI.MktIdxdGet('20160630', ticker = '000300', field = 'closeIndex').values[0][0]
# print fut,curr
```
