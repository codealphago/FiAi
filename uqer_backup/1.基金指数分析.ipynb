```python
def index_ma(indexid):
    trade_days = report.trade_pre_n_d(n=90)
    df=DataAPI.MktIdxdGet(ticker=indexid,beginDate=trade_days[0],endDate=trade_days[-1],field="tradeDate,lowestIndex,closeIndex,turnoverValue,secShortName",pandas="1")
    lowest=np.array(df['lowestIndex'])
    vol=np.array(df['turnoverValue'])
    mytitle = df.secShortName[0]
    #MA_Type: 0=SMA, 1=EMA, 2=WMA, 3=DEMA, 4=TEMA, 5=TRIMA, 6=KAMA, 7=MAMA, 8=T3 (Default=SMA)
    EMA20 = talib.MA(lowest,20,matype=0)
    VMA20 = talib.MA(vol,20,matype=1)
    # 20均线，量价齐升开始买入
    df['EMA20'] = EMA20
    df['VMA20'] = VMA20
    df['eFlag'] = df.lowestIndex - df.EMA20 
    df['vFlag'] = df.turnoverValue - df.VMA20
    df = df.dropna()
    df.index = df.tradeDate
    ax = df.plot(y=['lowestIndex','EMA20'], figsize=(18, 4),kind='line',title = unicode(mytitle, "utf-8") )
    report.fig_style(ax, [u'当日低价',u'价格均线'],'upper right')
    bx = df.plot(y=['turnoverValue','VMA20'], figsize=(18, 4),kind='line',title = unicode(mytitle, "utf-8") )
    report.fig_style(bx, [u'当日成交量',u'成交量均线'],'upper right') 
    return df
#买入条件: lowest[-1] >= EMA20[-1] and vol[-1] > VMA20[-1]:
#卖出条件: lowest[-1] < EMA20[-1] 
                              
```

```python
def index_mark(indexid):
    years = YearLocator()   # every year
    months = MonthLocator()  # every month
    yearsFmt = DateFormatter('%Y')

    color_map = {'LL':'r','LN':'y','SS':'b','SN':'g','XX':'k'}
    marker_map = {'LL':'^','LN':'*','SS':'s','SN':'d','XX':'.'}
    label_map = {'LL':'bull','LN':'rev','SS':'sell','SN':'SN','XX':'NOR'}
    
    trade_days = report.trade_pre_n_d(n=30)
    his = DataAPI.MktIdxdGet(ticker=indexid,beginDate=trade_days[0],endDate=trade_days[-1],field=u"tradeDate,highestIndex,lowestIndex,openIndex,preCloseIndex,turnoverVol,turnoverValue,CHGPct,closeIndex,ticker,secShortName",pandas="1")
    #his.index = his.tradeDate
    mytitle = his.secShortName.values[0]
    his['volMean'] = pd.rolling_mean(his.turnoverVol,5)
    his['priceMean'] =  pd.rolling_mean(his.closeIndex,5)
    his['CHGPctMean'] =  pd.rolling_mean(his.CHGPct,5)
    his= his.dropna()
    his = his.reset_index()
    size = len(his) 

    #初始标志   
    his['hidden_mark'] = 'XX'
 
    for x in range(1,size):
        pre = his.loc[x-1]
        curr = his.loc[x]
        volChg = abs(np.log(curr.turnoverVol) - np.log(pre.volMean))
        priceChg = np.log(curr.highestIndex) - np.log(curr.closeIndex)
        
        # 抛售标志(指数下跌 and 成交量 > 前日 and 成交量 > 前10日移动平均 )
        if curr.CHGPct < -0.002 and curr.turnoverVol > pre.turnoverVol  :#and curr.turnoverVol > curr.volMean:
            his.hidden_mark[x] = 'SS'
        # 滞涨行为(前期成交活跃,市场走高,之后遇到阻力,不能走高,上升幅度比不上前一两天)
        if abs(curr.CHGPct) < 0.003  and priceChg > 0.01 and volChg < 0.1 and curr.CHGPctMean > 0.005  :
            his.hidden_mark[x] = 'SN' 
        # 反弹开始信号
        if curr.CHGPct > 0.01 and curr.turnoverVol > pre.turnoverVol  and curr.turnoverVol > curr.volMean and priceChg < 0.003:
            his.hidden_mark[x] = 'LL'
        # 牛市信号(都是以一个反转日开始,大多数反转都发生在大盘试探性反弹的7天里的第4天-7天)
        

    # 绘图.
    plt.figure(figsize=(18, 4)) 
    for i in range(len(his)):
        flag =  his.hidden_mark[i] 
        plt.plot( his.index.values[i] , his.closeIndex.values[i],marker =marker_map[flag],lw=1 ,color=color_map[flag]) 
        plt.legend()
        plt.grid(1) 
    
    
    # ax = his.plot(y=['closeIndex'], figsize=(18, 4),title = unicode(mytitle, "utf-8"),colors='r',linestyle='',marker="d")
    # report.fig_style(ax, [u'收盘价'],'upper right')
    return his

index_mark(index_list[posi]).tail(2)
```
